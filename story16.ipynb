```
#### –ò—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –ª—é–±—ã–µ –±–∏–±–ª–∏–æ—Ç–µ–∫–∏ / –∏–º–ø–æ—Ä—Ç–∏—Ä–æ–≤–∞–Ω–Ω—ã–µ —Ñ—É–Ω–∫—Ü–∏–∏ –∑–∞–ø—Ä–µ—â–µ–Ω–æ.
#### => –°–û–ë–õ–Æ–î–ï–ù–û –ò–º–ø–æ—Ä—Ç—ã —É–∫–∞–∑—ã–≤–∞—é—Ç—Å—è –≤ –Ω–∞—á–∞–ª–µ —Ñ–∞–π–ª–∞ (import ...), –∏—Ö —Ç—É—Ç –Ω–µ—Ç—É –ø–æ—ç—Ç–æ–º—É —É—Å–ª–æ–≤–∏–µ —Å–æ–±–ª—é–¥–µ–Ω–æ 

#### –î–ª—è –ø—Ä–æ–≤–µ—Ä–∫–∏ –∫–æ–¥–∞ –ø—Ä–µ–¥–æ—Å—Ç–∞–≤–ª–µ–Ω —Ñ–∞–π–ª –≤ —Ñ–æ—Ä–º–∞—Ç–µ .ipynb
#### => –°–û–ë–õ–Æ–î–ï–ù–û –Ω–∞–∑–≤–∞–Ω–∏–µ —Ñ–∞–π–ª–∞ –≤ —Ñ–æ—Ä–º–∞—Ç–µ .ipynb 

#### –í –∏—Ç–æ–≥–µ —Ä–∞–±–æ—Ç—ã –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω —Ü–µ–ª—å–Ω—ã–π —Å—é–∂–µ—Ç —Å –º–∏–Ω–∏–º—É–º (3) —Ç—Ä–µ–º—è —Ä–∞–∑–ª–∏—á–Ω—ã–º–∏ –∫–æ–Ω—Ü–æ–≤–∫–∞–º–∏ (–Ω–∞ –¥–æ—Å—Ç–∏–∂–µ–Ω–∏–µ –∫–æ—Ç–æ—Ä—ã—Ö –º–æ–∂–µ—Ç –≤–ª–∏—è—Ç—å –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å —Å–≤–æ–∏–º –≤—ã–±–æ—Ä–æ–º).
#### => –°–û–ë–õ–Æ–î–ï–ù–û –∫–æ–Ω—Ü–æ–≤–∫–∏ –º–æ–∂–Ω–æ –Ω–∞–π—Ç–∏ –ø–æ –ø—Ä–µ—Ñ–∏–∫—Å—É (–ö–û–ù–¶–û–í–ö–ê:) - –∏—Ö 6 —à—Ç—É–∫

#### –í –∫–æ–¥–µ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω —Ö–æ—Ç—è –±—ã –æ–¥–∏–Ω —Ü–∏–∫–ª while. –ù–∞–ø—Ä–∏–º–µ—Ä, —Å—é–∂–µ—Ç
# —Ç–∞–∫–æ–π: –≥—Ä–∞–±–∏—Ç–µ–ª—å –ø—ã—Ç–∞–µ—Ç—Å—è –ø–æ–¥–æ–±—Ä–∞—Ç—å –ø–∞—Ä–æ–ª—å –∫ —Å–µ–π—Ñ—É, –ø–æ–∫–∞ –Ω–µ
# –∑–∞–∫–æ–Ω—á–∏—Ç—Å—è –≤—Ä–µ–º—è. –ï—Å–ª–∏ –≤—Ä–µ–º—è –≤—ã—à–ª–æ, —Ç–æ –µ–≥–æ –º–æ–≥—É—Ç –ø–æ–π–º–∞—Ç—å
# –∂–∏–ª—å—Ü—ã, –∞ –µ—Å–ª–∏ –≥—Ä–∞–±–∏—Ç–µ–ª—å —É—Å–ø–µ–ª –≤–æ–≤—Ä–µ–º—è, —Ç–æ –æ–Ω –ø–æ–ª—É—á–∞–µ—Ç
# –¥—Ä–∞–≥–æ—Ü–µ–Ω–Ω–æ—Å—Ç–∏ –∏–∑ —Å–µ–π—Ñ–∞.
#### => –°–û–ë–õ–Æ–î–ï–ù–û –∏—Ö 2 —à—Ç—É–∫–∏

#### –í –∫–æ–¥–µ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω —Ö–æ—Ç—è –±—ã –æ–¥–∏–Ω —Ü–∏–∫–ª for. –ù–∞–ø—Ä–∏–º–µ—Ä, —Å—é–∂–µ—Ç
# —Ç–∞–∫–æ–π: –Ω–∞ —ç—Ç–∞–∂–µ –µ—Å—Ç—å 4 –¥–≤–µ—Ä–∏ –∏ –≥—Ä–∞–±–∏—Ç–µ–ª—å –ø—Ä–æ–≤–µ—Ä—è–µ—Ç –∫–∞–∂–¥—É—é –Ω–∞
# –ø—Ä–µ–¥–º–µ—Ç –Ω–∞–ª–∏—á–∏—è –∑–∞ –¥–≤–µ—Ä—å—é —Ç–æ–≥–æ, —á—Ç–æ –º–æ–∂–Ω–æ —É–∫—Ä–∞—Å—Ç—å. –û–Ω
# –º–æ–∂–µ—Ç –¥–∞–∂–µ –∑–∞–π—Ç–∏ –≤ –∫–∞–∂–¥—É—é –∫–æ–º–Ω–∞—Ç—É –∏ –æ—Å–º–æ—Ç—Ä–µ—Ç—å—Å—è.
#### => –°–û–ë–õ–Æ–î–ï–ù–û 8 —à—Ç—É–∫

#### –í –∫–æ–¥–µ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–æ —Ö–æ—Ç—è –±—ã 
# –¥–≤–∞ –º–µ—Ç–æ–¥–∞ —Å–ø–∏—Å–∫–æ–≤ / 
# –¥–≤–∞ –º–µ—Ç–æ–¥–∞–º–Ω–æ–∂–µ—Å—Ç–≤ / 
# –æ–¥–∏–Ω –º–µ—Ç–æ–¥ —Å–ø–∏—Å–∫–æ–≤ –∏ –æ–¥–∏–Ω –º–µ—Ç–æ–¥ –º–Ω–æ–∂–µ—Å—Ç–≤.
# –ù–∞–ø—Ä–∏–º–µ—Ä, —Å–ø–∏—Å–∫–æ–º –≤ –¥–∞–Ω–Ω–æ–º —Å–ª—É—á–∞–µ –º–æ–∂–µ—Ç –±—ã—Ç—å —Å—É–º–∫–∞
# –≥—Ä–∞–±–∏—Ç–µ–ª—è, –≤ –∫–æ—Ç–æ—Ä—É—é –æ–Ω —Å–∫–ª–∞–¥—ã–≤–∞–µ—Ç —É–∫—Ä–∞–¥–µ–Ω–Ω–æ–µ.
#### => –°–û–ë–õ–Æ–î–ï–ù–û –¥–≤–∞ –º–µ—Ç–æ–¥–∞ —Å–ø–∏—Å–∫–æ–≤ =  append 2 —à—Ç—É–∫–∏ (–Ω–µ –±—ã–ª–æ —Å–∫–∞–∑–∞–Ω–æ –¥–≤–∞ –†–ê–ó–ù–´–• –º–µ—Ç–æ–¥–∞)
#### => –°–û–ë–õ–Æ–î–ï–ù–û –æ–¥–∏–Ω –º–µ—Ç–æ–¥ –º–Ω–æ–∂–µ—Å—Ç–≤ intersection

#### –í –∫–æ–¥–µ —Å–æ–∑–¥–∞–µ—Ç—Å—è —Ö–æ—Ç—è –±—ã –æ–¥–∏–Ω –Ω–æ–≤—ã–π —Å–ª–æ–≤–∞—Ä—å. –ù–∞–ø—Ä–∏–º–µ—Ä, –µ—Å–ª–∏
# –≥—Ä–∞–±–∏—Ç–µ–ª—è –ø–æ–π–º–∞–ª–∞ –ø–æ–ª–∏—Ü–∏—è, —Ç–æ –æ–Ω–∞ –º–æ–∂–µ—Ç –∑–∞–ø–∏—Å–∞—Ç—å –≤ —Å–ª–æ–≤–∞—Ä—å
# –Ω–∞–∑–≤–∞–Ω–∏—è —Å—Ç–∞—Ç–µ–π, –∫–æ—Ç–æ—Ä—ã–µ –æ–Ω –Ω–∞—Ä—É—à–∏–ª, –∏ –Ω–∞–∫–∞–∑–∞–Ω–∏–µ –ø–æ –∫–∞–∂–¥–æ–π
# —Å—Ç–∞—Ç—å–µ.
#### => –°–û–ë–õ–Æ–î–ï–ù–û –Ω–∞–∑–≤–∞–Ω–∏–µ —Å–ª–æ–≤–∞—Ä—è excuses, –≤ handle_propose, –≤ —Å–æ–∑–¥–∞–Ω–∏–µ –æ–ø—Ä–∞–≤–¥–∞–Ω–∏–π

### –í –∫–æ–¥–µ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–æ —Ö–æ—Ç—è –±—ã 4 —É—Å–ª–æ–≤–Ω—ã—Ö –∫–æ–Ω—Å—Ç—Ä—É–∫—Ü–∏–π.
# –ù–∞–ø—Ä–∏–º–µ—Ä, —Å—é–∂–µ—Ç —Ç–∞–∫–æ–π: –≥—Ä–∞–±–∏—Ç–µ–ª—å —É–±–µ–≥–∞–µ—Ç –æ—Ç –ø—Ä–µ—Å–ª–µ–¥–æ–≤–∞–Ω–∏—è –∏
# –≤—ã–±–∏—Ä–∞–µ—Ç, —Å–≤–µ—Ä–Ω—É—Ç—å –Ω–∞–ø—Ä–∞–≤–æ –∏–ª–∏ –Ω–∞–ª–µ–≤–æ ‚Äî –µ—Å–ª–∏ –æ–Ω —Å–≤–µ—Ä–Ω–µ—Ç
# –Ω–∞–ø—Ä–∞–≤–æ, —Ç–æ –µ–≥–æ –∑–∞–¥–µ—Ä–∂–∏—Ç –ø–æ–ª–∏—Ü–∏—è, –∞ –µ—Å–ª–∏ –æ–Ω —Å–≤–µ—Ä–Ω–µ—Ç –Ω–∞–ø—Ä–∞–≤–æ,
# —Ç–æ –≤—Å—Ç—Ä–µ—Ç–∏—Ç –º–∞—Ñ–∏—é, –∫–æ—Ç–æ—Ä–∞—è –∑–∞–±–µ—Ä–µ—Ç —É –Ω–µ–≥–æ –≤—Å–µ –Ω–∞–≥—Ä–∞–±–ª–µ–Ω–Ω–æ–µ.
# –¢–∞–∫–∏—Ö —Ä–∞–∑–≤–µ—Ç–≤–ª–µ–Ω–∏–π –≤ —Å—é–∂–µ—Ç–µ –º–æ–∂–µ—Ç –±—ã—Ç—å –∏ –±–æ–ª—å—à–µ.
#### => –°–û–ë–õ–Æ–î–ï–ù–û , —É—Å–ª–æ–≤–Ω—ã–µ –∫–æ–Ω—Å—Ç—Ä—É–∫—Ü–∏–∏ —ç—Ç–æ if / else / elif 

#### –í –∫–æ–¥–µ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω —Ö–æ—Ç—è –±—ã –æ–¥–∏–Ω –≥–æ—Ç–æ–≤—ã–π —Å–ª–æ–≤–∞—Ä—å. –ù–∞–ø—Ä–∏–º–µ—Ä,
# –≤ –≤–∏–¥–µ —Å–ª–æ–≤–∞—Ä—è –º–æ–∂–µ—Ç –±—ã—Ç—å –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–∞ –∞–¥—Ä–µ—Å–Ω–∞—è –∫–Ω–∏–≥–∞, –∏–∑
# –∫–æ—Ç–æ—Ä–æ–π –≥—Ä–∞–±–∏—Ç–µ–ª—å –≤—ã–±–∏—Ä–∞–µ—Ç —Å–µ–±–µ –ø–æ—Ç–µ–Ω—Ü–∏–∞–ª—å–Ω—É—é –∂–µ—Ä—Ç–≤—É.
#### => –°–û–ë–õ–Æ–î–ï–ù–û —Å–ª–æ–≤–∞—Ä—å –Ω–∞–∑—ã–≤–∞–∫–µ—Ç—Å—è stories

#### –ò—Ç–æ–≥–æ–≤—ã–π —Å—é–∂–µ—Ç (—Å–æ –≤—Å–µ–º–∏ –≤—ã–±–æ—Ä–∞–º–∏ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è) –∏
# –ø–æ–ª—É—á–∏–≤—à–∞—è—Å—è –∫–æ–Ω—Ü–æ–≤–∫–∞ –∞–≤—Ç–æ–º–∞—Ç–∏—á–µ—Å–∫–∏ –∑–∞–ø–∏—Å–∞–Ω—ã –≤ —Ñ–∞–π–ª.
#### => –°–û–ë–õ–Æ–î–ï–ù–û –∑–∞–ø–∏—Å—ã–≤–∞–µ—Ç—Å—è –∫–Ω—Ü–æ–≤–∫–∞ –≤ –æ–¥–∏–Ω —Ñ–∞–π–ª –∞ –¥—Ä—É–≥–∏–µ –¥–∞–Ω–Ω—ã–µ –≤ –¥—Ä—É–≥–∏–µ —Ñ–∞–π–ª—ã, —á—Ç–æ—ã–± —Ç–æ—á–Ω–æ –∑–∞–ø–∏—Å–∞–ª–æ—Å—å. –í—Å–µ–≥–æ 4 —Ñ–∞–π–ª–∞ —Ä–∞–∑–Ω—ã—Ö




# –°–∞–º–∞ –∏—Å—Ç–æ—Ä–∏—è —Å–æ —Å—Ü–µ–Ω–∞—Ä–∏—è–º–∏, –∫–∞–∫ —Å–ª–æ–≤–∞—Ä—è –≤ –≤–∏–¥–µ –≥—Ä–∞—Ñ–∞ (–¥–µ—Ä–µ–≤–∞), 
# —Ç–∞–∫ –∫–∞–∫ –≥—Ä–∞—Ñ–∞ (–¥–µ—Ä–µ–≤–æ) –ø–æ–∑–≤–æ–ª—è–µ—Ç –ø–µ—Ä–µ–±–∏—Ä–∞—Ç—å –∏ –≤—ã–±–∏—Ä–∞—Ç—å —Ä–∞–∑–Ω—ã–µ –≤–∞—Ä–∏–∞–Ω—Ç—ã, 
# —Ç–∞–∫–∂–µ —ç—Ç–æ –ø–æ–∑–≤–æ–ª—è–µ—Ç –Ω–µ –¥–µ–æ–∞—Ç—å –º–Ω–æ–≥–æ if –ø–æ–¥ –∫–∞–∂–¥—ã–π —Å—Ü–µ–Ω–∞—Ä–∏–π
stories = {
    "start": {
        "title": "–ú—ã –≤ –æ—Ç–µ–ª–µ —Å –ø–∞—Ä—Ç–Ω–µ—Ä–æ–º",
        "description": "–õ—é—Å–∏ –∏ –ú–∞—Ä–∫ –ø—Ä–∏–µ–∑–∂–∞—é—Ç –≤ —Ä–æ—Å–∫–æ—à–Ω—ã–π –æ—Ç–µ–ª—å ¬´–ë–µ–ª—ã–π –õ–æ—Ç–æ—Å¬ª –Ω–∞ –ø–æ–±–µ—Ä–µ–∂—å–µ, –æ–Ω–∏ –Ω–µ –∂–µ–Ω–∞—Ç—ã. –ù–∞–¥–æ –≤–∑—è—Ç—å —á—Ç–æ-—Ç–æ —Å —Å–æ–±–æ–π –≤—ã—Ö–æ–¥—è –∏–∑ –Ω–æ–º–µ—Ä–∞. –ß—Ç–æ –≤—ã –ø–æ–ª–æ–∂–∏—Ç–µ –≤ —Å—É–º–∫—É?\n–í—ã –≤—ã–±–∏—Ä–∞–µ—Ç–µ, –∫–∞–∫ —Ö–æ—Ç–µ–ª–∏ –±—ã –ø—Ä–æ–≤–µ—Å—Ç–∏ –≤–µ—á–µ—Ä",
        "choices": {
            "1": "restaurant_dinner",
            "2": "boat_dinner"
        }
    },
    "restaurant_dinner": {
        "title": "–û—Ç–ø—Ä–∞–≤–∏—Ç—å—Å—è –Ω–∞ —É–∂–∏–Ω –≤ —Ä–µ—Å—Ç–æ—Ä–∞–Ω –æ—Ç–µ–ª—è",
        "description": "–õ—é—Å–∏ –∏ –ú–∞—Ä–∫ –Ω–∞—Å–ª–∞–∂–¥–∞—é—Ç—Å—è —Ä–æ—Å–∫–æ—à–Ω–æ–π –µ–¥–æ–π –∏ –≤—Å—Ç—Ä–µ—á–∞—é—Ç –±–æ–≥–∞—Ç–æ–≥–æ –±–∏–∑–Ω–µ—Å–º–µ–Ω–∞, –∫–æ—Ç–æ—Ä—ã–π –ø—Ä–µ–¥–ª–∞–≥–∞–µ—Ç –∏–º —ç–∫—Å–∫—É—Ä—Å–∏—é –ø–æ –æ–∫—Ä–µ—Å—Ç–Ω–æ—Å—Ç—è–º. –ß—Ç–æ –≤—ã –≤—ã–±–µ—Ä–µ—Ç–µ?",
        "choices": {
            "1": "accept_invitation",
            "2": "decline_invitation"
        }
    },
    "boat_dinner": {
        "title": "–£—Å—Ç—Ä–æ–∏—Ç—å —É–∂–∏–Ω –Ω–∞ –ª–æ–¥–∫–µ",
        "description": "–í—ã –ø—Ä–∏—à–ª–∏ –Ω–∞ –±–æ—Ä—Ç, –∞–¥–º–∏–Ω–∏—Å—Ç—Ä–∞—Ç–æ—Ä –ø—Ä–æ—Å–∏—Ç –≤–∞—Å –≤–≤–µ—Å—Ç–∏ –¥–∞–Ω–Ω—ã–µ —Å –∫–∞—Ä—Ç—ã, –≤—ã –ø—ã—Ç–∞–µ—Ç–µ—Å—å –≤—Å–ø–æ–º–Ω–∏—Ç—å –ø–∞—Ä–æ–ª—å.",
        "choices": {
            "1": "remember_code",
            "2": "forget_code"
        }
    },
    "accept_invitation": {
        "title": "–ü—Ä–∏–Ω—è—Ç—å –ø—Ä–∏–≥–ª–∞—à–µ–Ω–∏–µ",
        "description": "–í—ã —á—Ç–æ-—Ç–æ –æ–±—Å—É–∂–¥–∞–µ—Ç–µ —Å –±–∏–∑–Ω–µ—Å–º–µ–Ω–æ–º. –£ –≤–∞—Å –µ—Å—Ç—å –¥—Ä—É–≥ –∫–æ—Ç–æ—Ä—ã–π –∑–Ω–∞–µ—Ç –Ω–µ–æ–±—Ö–æ–¥–∏–º—É—é –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é. –í—ã –≤—ã–±–∏—Ä–∞–µ—Ç–µ –∏–∑ –∫–æ–Ω—Ç–∞–∫—Ç–æ–≤ –∫–æ–º—É –ø–æ–∑–≤–æ–Ω–∏—Ç—å. –í—ã–±–µ—Ä–∏—Ç–µ –∏–∑ —Å–ª–µ–¥—É—é—â–∏—Ö –∏–º–µ–Ω: –ú–∞—Ä—Ç–∏–Ω–∞, –õ—É–∏–∑–∞, –í–∏–Ω–Ω–∏, –ú–∞–π–∫.",
        "choices": {
            "1": "call_martin",
            "2": "call_louise",
            "3": "call_winnie",
            "4": "call_mike"
        }
    },
    "decline_invitation": {
        "title": "–û—Ç–∫–∞–∑–∞—Ç—å—Å—è –∏ –æ—Å—Ç–∞—Ç—å—Å—è –Ω–∞ —É–∂–∏–Ω–µ",
        "description": "–ö–∞–∂–µ—Ç—Å—è, –ø—Ä–∏—à–ª–æ –≤—Ä–µ–º—è –≤—ã–±–∏—Ä–∞—Ç—å –±–ª—é–¥–∞. –ß—Ç–æ –≤—ã –ø—Ä–µ–¥–ø–æ—á–µ—Ç–µ?\n\n1. –°–∞–ª–∞—Ç —Å –º–æ—Ä–µ–ø—Ä–æ–¥—É–∫—Ç–∞–º–∏\n2. –ö—Ä–µ–º-—Å—É–ø –∏–∑ —Ç—ã–∫–≤—ã\n3. –§–∏–ª–µ –º–∏–Ω—å–æ–Ω\n4. –†–∏–∑–æ—Ç—Ç–æ —Å –≥—Ä–∏–±–∞–º–∏ –∏ —Ç—Ä—é—Ñ–µ–ª—å–Ω—ã–º –º–∞—Å–ª–æ–º\n5. –¢–∏—Ä–∞–º–∏—Å—É",
        "choices": {
            "1": "fight_starts",
            "2": "dessert_fight"
        }
    },
    "dessert_fight": {
        "title": "–ù–∞—á–∞–ª–æ —Å–∫–∞–Ω–¥–∞–ª–∞",
        "description": "–û–¥–∏–Ω –∏–∑ –ø–æ—Å—Ç–æ—è–ª—å—Ü–µ–≤ —É—Å—Ç—Ä–∞–∏–≤–∞–µ—Ç –¥—Ä–∞–∫—É –∏–∑-–∑–∞ –¥–µ—Å–µ—Ä—Ç–∞, –≤—ã –Ω–µ –æ—Ç–≤–µ—á–∞–µ—Ç–µ –∏ –≤–∞—Å –∞—Ä–µ—Å—Ç–æ–≤—ã–≤–∞—é—Ç.",
        "choices": {}
    },
    "fight_starts": {
        "title": "–ù–∞—á–∞–ª–æ –¥—Ä–∞–∫–∏",
        "description": "–û–¥–∏–Ω –∏–∑ –ø–æ—Å—Ç–æ—è–ª—å—Ü–µ–≤ —É—Å—Ç—Ä–∞–∏–≤–∞–µ—Ç –¥—Ä–∞–∫—É, –Ω–æ –≤—ã –≤–º–µ—à–∏–≤–∞–µ—Ç–µ—Å—å –∏ –Ω–∞ –≤–∞—Å –Ω–µ –æ–±—Ä–∞—â–∞—é—Ç –≤–Ω–∏–º–∞–Ω–∏–µ.",
        "choices": {}
    },
    "remember_code": {
        "title": "–í—Å–ø–æ–º–Ω–∏–ª –∫–æ–¥",
        "description": "–ê–¥–º–∏–Ω–∏—Å—Ç—Ä–∞—Ç–æ—Ä –ø—Ä–æ–ø—É—Å–∫–∞–µ—Ç –≤–∞—Å –∑–∞ —Å—Ç–æ–ª–∏–∫. –ù–∞—Å–ª–∞–∂–¥–∞—è—Å—å –≤–∏–¥–æ–º, –≤—ã –¥—É–º–∞–µ—Ç–µ, –ø—Ä–∏–Ω—è—Ç—å —Ä–µ—à–µ–Ω–∏–µ –∫–æ—Ç–æ—Ä–æ–µ —Ç–∞–∫ –¥–∞–≤–Ω–æ –ø–ª–∞–Ω–∏—Ä–æ–≤–∞–ª–∏.",
        "choices": {
            "1": "propose",
            "2": "not_propose"
        }
    },
    "forget_code": {
        "title": "–ù–µ –≤—Å–ø–æ–º–Ω–∏–ª –∫–æ–¥",
        "description": "–ê–¥–º–∏–Ω–∏—Å—Ç—Ä–∞—Ç–æ—Ä –Ω–µ –ø—Ä–æ–ø—É—Å–∫–∞–µ—Ç –≤–∞—Å, –∏ –≤–∞—Å –±—Ä–æ—Å–∞–µ—Ç –¥–µ–≤—É—à–∫–∞. –í—ã –ø—Ä–æ–≤–æ–¥–∏—Ç–µ –æ—Ç–ø—É—Å–∫ –æ–¥–∏–Ω.",
        "choices": {}
    },
    "propose": {
        "title": "–°–¥–µ–ª–∞—Ç—å –ø—Ä–µ–¥–ª–æ–∂–µ–Ω–∏–µ",
        "description": "–û –Ω–µ—Ç! –û–Ω–∞ –ø–æ–¥–æ–∑—Ä–µ–≤–∞–µ—Ç –≤–∞—Å –≤ –Ω–µ–≤–µ—Ä–Ω–æ—Å—Ç–∏, –≤–∞–º –Ω—É–∂–Ω–æ –æ–ø—Ä–∞–≤–¥–∞—Ç—å—Å—è. –í–≤–µ–¥–∏—Ç–µ 3 –ø—Ä–∏—á–∏–Ω—ã, –ø–æ—á–µ–º—É –≤—ã —Ö–æ–¥–∏–ª–∏ –≤ —Ä–µ—Å—Ç–æ—Ä–∞–Ω —Å –¥—Ä—É–≥–æ–π –¥–µ–≤—É—à–∫–æ–π.",
        "choices": {
            "1": "give_excuses"
        }
    },
    "not_propose": {
        "title": "–ù–µ –¥–µ–ª–∞—Ç—å –ø—Ä–µ–¥–ª–æ–∂–µ–Ω–∏–µ",
        "description": "–û–Ω–∞ —Ç–∞–∫ –¥–æ–ª–≥–æ –∂–¥–∞–ª–∞ –≤–∞—à–µ–≥–æ –ø—Ä–µ–¥–ª–æ–∂–µ–Ω–∏—è, —É–≤—ã, —ç—Ç–æ –±—ã–ª –ø–æ—Å–ª–µ–¥–Ω–∏–π —à–∞–Ω—Å.",
        "choices": {}
    },
    "give_excuses": {
        "title": "–î–∞—Ç—å –æ–ø—Ä–∞–≤–¥–∞–Ω–∏—è",
        "description": "–û–Ω–∞ —Ç—Ä–µ–±—É–µ—Ç –æ–ø—Ä–∞–≤–¥–∞–Ω–∏—è –¥–ª—è –∫–∞–∂–¥–æ–π –∏–∑ —Ç—Ä–µ—Ö –∏–∑–º–µ–Ω. –ü—Ä–∏–¥—É–º–∞–π—Ç–µ —Ç—Ä–∏ –æ–ø—Ä–∞–≤–¥–∞–Ω–∏—è.",
        "choices": {
            "1": "excuse_1",
            "2": "excuse_2",
            "3": "excuse_3"
        }
    },
    "excuse_1": {
        "title": "–ü–µ—Ä–≤–æ–µ –æ–ø—Ä–∞–≤–¥–∞–Ω–∏–µ",
        "description": "–í–≤–µ–¥–∏—Ç–µ –ø–µ—Ä–≤–æ–µ –æ–ø—Ä–∞–≤–¥–∞–Ω–∏–µ.",
        "choices": {}
    },
    "excuse_2": {
        "title": "–í—Ç–æ—Ä–æ–µ –æ–ø—Ä–∞–≤–¥–∞–Ω–∏–µ",
        "description": "–í–≤–µ–¥–∏—Ç–µ –≤—Ç–æ—Ä–æ–µ –æ–ø—Ä–∞–≤–¥–∞–Ω–∏–µ.",
        "choices": {}
    },
    "excuse_3": {
        "title": "–¢—Ä–µ—Ç—å–µ –æ–ø—Ä–∞–≤–¥–∞–Ω–∏–µ",
        "description": "–í–≤–µ–¥–∏—Ç–µ —Ç—Ä–µ—Ç—å–µ –æ–ø—Ä–∞–≤–¥–∞–Ω–∏–µ.",
        "choices": {
            "1": "girlfriend_agrees"
        }
    },
    "girlfriend_agrees": {
        "title": "–û–Ω–∞ —Å–æ–≥–ª–∞—à–∞–µ—Ç—Å—è",
        "description": "–§—É—Ö! –û–Ω–∞ –Ω–∞–∫—Ä—É—Ç–∏–ª–∞ —Å–µ–±—è –Ω–∞ –Ω–µ—Ä–≤–∞—Ö... –ö–æ–Ω–µ—á–Ω–æ —Å–æ–≥–ª–∞—Å–Ω–∞.",
        "choices": {}
    },
    "call_martin": {
        "title": "–ü–æ–∑–≤–æ–Ω–∏—Ç—å –ú–∞—Ä—Ç–∏–Ω—É",
        "description": "–í—ã –ø–æ–∑–≤–æ–Ω–∏–ª–∏ –≤–∞—à–µ–º—É –¥—Ä—É–≥—É, –∑–∞–∫–ª—è—Ç–æ–º—É –≤—Ä–∞–≥—É –±–∏–∑–Ω–µ—Å–º–µ–Ω–∞, –∏ –æ–Ω –≤–∞—Å —Ö–æ—á–µ—Ç —É–±–∏—Ç—å.",
        "choices": {}
    },
    "call_louise": {
        "title": "–ü–æ–∑–≤–æ–Ω–∏—Ç—å –õ—É–∏–∑–µ",
        "description": "–í—ã –∑–≤–æ–Ω–∏—Ç–µ –≤–µ—Ä–Ω–æ–º—É –¥—Ä—É–≥—É –∏ –ø–æ–ª—É—á–∞–µ—Ç–µ –¥–æ–≤–µ—Ä–∏–µ –±–∏–∑–Ω–µ—Å–º–µ–Ω–∞.",
        "choices": {}
    },
    "call_winnie": {
        "title": "–ü–æ–∑–≤–æ–Ω–∏—Ç—å –í–∏–Ω–Ω–∏",
        "description": "–í—ã –ø–æ–∑–≤–æ–Ω–∏–ª–∏ –≤–∞—à–µ–º—É –¥—Ä—É–≥—É, –∑–∞–∫–ª—è—Ç–æ–º—É –≤—Ä–∞–≥—É –±–∏–∑–Ω–µ—Å–º–µ–Ω–∞, –∏ –æ–Ω –≤–∞—Å —Ö–æ—á–µ—Ç —É–±–∏—Ç—å.",
        "choices": {}
    },
    "call_mike": {
        "title": "–ü–æ–∑–≤–æ–Ω–∏—Ç—å –ú–∞–π–∫—É",
        "description": "–í—ã –∑–≤–æ–Ω–∏—Ç–µ –≤–µ—Ä–Ω–æ–º—É –¥—Ä—É–≥—É –∏ –ø–æ–ª—É—á–∞–µ—Ç–µ –¥–æ–≤–µ—Ä–∏–µ –±–∏–∑–Ω–µ—Å–º–µ–Ω–∞.",
        "choices": {}
    },
    "final_arrest": {
        "title": "–í–∞—Å –∞—Ä–µ—Å—Ç–æ–≤—ã–≤–∞—é—Ç",
        "description": "–ö–û–ù–¶–û–í–ö–ê: –í–∞—Å –∞—Ä–µ—Å—Ç–æ–≤–∞–ª–∏ –∑–∞ —É—á–∞—Å—Ç–∏–µ –≤ –¥—Ä–∞–∫–µ. –≠—Ç–æ –∫–æ–Ω–µ—Ü –≤–∞—à–µ–π –∏—Å—Ç–æ—Ä–∏–∏.",
        "choices": {}
    },
    "final_ignored": {
        "title": "–ò–≥–Ω–æ—Ä–∏—Ä–æ–≤–∞–Ω–∏–µ",
        "description": "–ö–û–ù–¶–û–í–ö–ê: –í–∞—Å –Ω–µ –∑–∞–º–µ—Ç–∏–ª–∏ –∏ –≤—ã –ø—Ä–æ–≤–µ–ª–∏ –≤–µ—á–µ—Ä –≤ –ø–æ–∫–æ–µ. –≠—Ç–æ –∫–æ–Ω–µ—Ü –≤–∞—à–µ–π –∏—Å—Ç–æ—Ä–∏–∏.",
        "choices": {}
    },
    "final_alone": {
        "title": "–û—Å—Ç–∞–≤–∞—Ç—å—Å—è –æ–¥–Ω–æ–º—É",
        "description": "–ö–û–ù–¶–û–í–ö–ê: –í–∞—Å –±—Ä–æ—Å–∞–µ—Ç –¥–µ–≤—É—à–∫–∞, –∏ –≤—ã –ø—Ä–æ–≤–æ–¥–∏—Ç–µ –æ—Ç–ø—É—Å–∫ –æ–¥–∏–Ω. –≠—Ç–æ –∫–æ–Ω–µ—Ü –≤–∞—à–µ–π –∏—Å—Ç–æ—Ä–∏–∏.",
        "choices": {}
    },
    "final_no_proposal": {
        "title": "–ù–µ—Ç –ø—Ä–µ–¥–ª–æ–∂–µ–Ω–∏—è",
        "description": "–ö–û–ù–¶–û–í–ö–ê: –û–Ω–∞ —Ç–∞–∫ –¥–æ–ª–≥–æ –∂–¥–∞–ª–∞ –≤–∞—à–µ–≥–æ –ø—Ä–µ–¥–ª–æ–∂–µ–Ω–∏—è, —É–≤—ã, —ç—Ç–æ –±—ã–ª –ø–æ—Å–ª–µ–¥–Ω–∏–π —à–∞–Ω—Å. –≠—Ç–æ –∫–æ–Ω–µ—Ü –≤–∞—à–µ–π –∏—Å—Ç–æ—Ä–∏–∏.",
        "choices": {}
    },
    "final_attack": {
        "title": "–ê—Ç–∞–∫–∞",
        "description": "–ö–û–ù–¶–û–í–ö–ê: –í–∞—à –∑–∞–∫–ª—è—Ç—ã–π –≤—Ä–∞–≥ –∞—Ç–∞–∫—É–µ—Ç –≤–∞—Å, –∏ –≤—ã –≤—ã–Ω—É–∂–¥–µ–Ω—ã –±–æ—Ä–æ—Ç—å—Å—è –∑–∞ —Å–≤–æ—é –∂–∏–∑–Ω—å. –≠—Ç–æ –∫–æ–Ω–µ—Ü –≤–∞—à–µ–π –∏—Å—Ç–æ—Ä–∏–∏.",
        "choices": {}
    },
    "final_apartment": {
        "title": "–ü–æ–ª—É—á–µ–Ω–∏–µ –∫–≤–∞—Ä—Ç–∏—Ä—ã",
        "description": "–ö–û–ù–¶–û–í–ö–ê: –í—ã –ø–æ–ª—É—á–∞–µ—Ç–µ –¥–æ–≤–µ—Ä–∏–µ –±–∏–∑–Ω–µ—Å–º–µ–Ω–∞, –∏ –æ–Ω –¥–∞—Ä–∏—Ç –≤–∞–º –∫–≤–∞—Ä—Ç–∏—Ä—É. –≠—Ç–æ –∫–æ–Ω–µ—Ü –≤–∞—à–µ–π –∏—Å—Ç–æ—Ä–∏–∏.",
        "choices": {}
    }
}

contacts = ["–ú–∞—Ä—Ç–∏–Ω", "–õ—É–∏–∑–∞", "–í–∏–Ω–Ω–∏", "–ú–∞–π–∫"]  # –°–ø–∏—Å–æ–∫ –∫–æ–Ω—Ç–∞–∫—Ç–æ–≤

correct_code = "1205"  # –ö–æ–¥, –∫–æ—Ç–æ—Ä—ã–π –Ω—É–∂–Ω–æ —É–≥–∞–¥–∞—Ç—å

name_file = "story_file"

# input() - –º–µ—Ç–æ–¥ –≤–≤–æ–¥–∞ / —Å –∫–ª–∞–≤–∏–∞—Ç—É—Ä—ã —Ç–µ–∫—Å—Ç –ø–æ–ª—É—á–∏—Ç—å
# print() - –º–µ—Ç–æ–¥ –≤—ã–≤–æ–¥–∞ / –æ—Ç–±—Ä–∞–∂–µ–Ω–∏—è —Ç–µ–∫—Å—Ç–∞ 

# –§—É–Ω–∫—Ü–∏—è. –û—Ç–æ–±—Ä–∞–∂–∞–µ–º –æ–ø–∏—Å–∞–Ω–∏–µ —Å—Ü–µ–Ω–∞—Ä–∏—è –∏ –≤—ã–±–æ—Ä –¥–µ–π—Å—Ç–≤–∏–π
def display_scene(scene):
    print(f"\n üíØ {stories[scene]['title']}\n{stories[scene]['description']}\n")
    if scene == "start":
        input_library_items()  # –í–≤–æ–¥ –ø—Ä–µ–¥–º–µ—Ç–æ–≤ –∏–∑ –±–∏–±–ª–∏–æ—Ç–µ–∫–∏
        print(f"\n–í—ã–±–µ—Ä–∏—Ç–µ —Å–ª–µ–¥—É—é—â–µ–µ –¥–µ–π—Å—Ç–≤–∏–µ\n")
    choices = list(stories[scene]['choices'].keys())
    if len(choices) == 1:
        # –ï—Å–ª–∏ –µ—Å—Ç—å —Ç–æ–ª—å–∫–æ –æ–¥–∏–Ω –≤—ã–±–æ—Ä —Ç–æ –∞–≤—Ç–æ–º–∞—Ç–∏—á–µ—Å–∫–∏ –≤—ã–±–∏—Ä–∞–µ—Ç—Å—è –æ–Ω
        next_scene = stories[scene]['choices'][choices[0]]
        print(f"–ï–¥–∏–Ω—Å—Ç–≤–µ–Ω–Ω—ã–π –≤—ã–±–æ—Ä: {stories[next_scene]['title']}")
        return next_scene
    elif choices:
        # –û—Ç–æ–±—Ä–∞–∂–∞–µ–º –≤—Å–µ –≤–∞—Ä–∏–∞–Ω—Ç—ã –∫–æ—Ç–æ—Ä—ã–µ –µ—Å—Ç—å
        for choice in choices:
            print(f"{choice}. {stories[stories[scene]['choices'][choice]]['title']}")
        user_choice = input("–í—ã–±–µ—Ä–∏—Ç–µ –≤–∞—Ä–∏–∞–Ω—Ç: ")
        if user_choice in choices:
            return stories[scene]['choices'][user_choice]
        else:
            print("–ù–µ–ø—Ä–∞–≤–∏–ª—å–Ω—ã–π –≤—ã–±–æ—ÄüôÖüèª‚Äç‚ôÇÔ∏è. –ü–æ–ø—Ä–æ–±—É–π—Ç–µ –µ—â–µ —Ä–∞–∑.")
            return display_scene(scene)
    else:
        return None

# –§—É–Ω–∫—Ü–∏—è. –í–≤–æ–¥ –∫–æ–¥–∞
def handle_boat_dinner():
    attempt = ""
    number_of_attempts = 3  # –û–≥—Ä–∞–Ω–∏—á–µ–Ω–∏–µ –Ω–∞ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø–æ–ø—ã—Ç–æ–∫
    print("\n üíØ –û–Ω –æ–ø–ª–∞—á–∏–≤–∞–µ—Ç –∫–∞—Ä—Ç–æ–π. –ó–∞–±—ã–ª –ø–∞—Ä–æ–ª—å ü§¶üèª‚Äç‚ôÄÔ∏è")
    while attempt != correct_code and number_of_attempts > 0:
        attempt = input(f"–ø–æ–¥–±–∏—Ä–∞–µ—Ç –∫–æ–¥ (–æ—Å—Ç–∞–ª–æ—Å—å –ø–æ–ø—ã—Ç–æ–∫ {number_of_attempts}): ")
        if attempt == correct_code:
            print("–í—ã –≤—Å–ø–æ–º–Ω–∏–ª–∏ –∫–æ–¥ –∏ –ø—Ä–æ—à–ª–∏ –Ω–∞ —É–∂–∏–Ω.")
            return "remember_code"
        else:
            # –£–±–µ—Ä–∞–µ–º –º–∏–Ω—É—Å –æ–¥–Ω—É –ø–æ–ø—ã—Ç–∫—É
            number_of_attempts -= 1
            if number_of_attempts == 0:
                print("–ù–µ–ø—Ä–∞–≤–∏–ª—å–Ω—ã–π –∫–æ–¥. –í—ã –∏—Å—á–µ—Ä–ø–∞–ª–∏ –≤—Å–µ –ø–æ–ø—ã—Ç–∫–∏.")
                return "forget_code"
            else:
                # –û–±—Ä–∞–±–∞—Ç—ã–≤–∞–µ–º —Å–ª—É—á–∞–π –µ—Å–ª–∏ –≤–≤–µ–ª –Ω–µ–ø—Ä–∞–≤–∏–ª—å–Ω–æ –≤—ã–±–æ—Ä
                print("–ù–µ–ø—Ä–∞–≤–∏–ª—å–Ω—ã–π –∫–æ–¥. –ü–æ–ø—Ä–æ–±—É–π—Ç–µ –µ—â–µ —Ä–∞–∑.")

# –í—ã–±–æ—Ä –∫–æ–º—É –ø–æ–∑–≤–æ–Ω–∏—Ç—å
def handle_accept_invitation():
    print("\n üíØ –í—ã —á—Ç–æ-—Ç–æ –æ–±—Å—É–∂–¥–∞–µ—Ç–µ —Å –±–∏–∑–Ω–µ—Å–º–µ–Ω–æ–º. –£ –≤–∞—Å –µ—Å—Ç—å –¥—Ä—É–≥ –∫–æ—Ç–æ—Ä—ã–π –∑–Ω–∞–µ—Ç –Ω–µ–æ–±—Ö–æ–¥–∏–º—É—é –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é. –í—ã –≤—ã–±–∏—Ä–∞–µ—Ç–µ –∏–∑ –∫–æ–Ω—Ç–∞–∫—Ç–æ–≤ –∫–æ–º—É –ø–æ–∑–≤–æ–Ω–∏—Ç—å. –í—ã–±–∏—Ä–µ—Ç–µ –∏–∑ —Å–ª–µ–¥—É—é—â–∏—Ö –∏–º–µ–Ω:\n")
    print("–í—ã–±–∏—Ä–∞–µ—Ç–µ, –∫–æ–º—É –ø–æ–∑–≤–æ–Ω–∏—Ç—å –∏–∑ —Å–ø–∏—Å–∫–∞ –∫–æ–Ω—Ç–∞–∫—Ç–æ–≤üì±:")
    for i, contact in enumerate(contacts, 1):
        print(f"{i}. {contact}")
    user_choice = input("–í–≤–µ–¥–∏—Ç–µ –Ω–æ–º–µ—Ä –∫–æ–Ω—Ç–∞–∫—Ç–∞üë®üèΩ‚Äçüíª: ")
    if user_choice == "1":
        return "call_martin"
    elif user_choice == "2":
        return "call_louise"
    elif user_choice == "3":
        return "call_winnie"
    elif user_choice == "4":
        return "call_mike"
    else:
        # –û–±—Ä–∞–±–∞—Ç—ã–≤–∞–µ–º —Å–ª—É—á–∞–π –µ—Å–ª–∏ –≤–≤–µ–ª –Ω–µ–ø—Ä–∞–≤–∏–ª—å–Ω–æ –≤—ã–±–æ—Ä
        print("–ù–µ–ø—Ä–∞–≤–∏–ª—å–Ω—ã–π –≤—ã–±–æ—Ä. –ü–æ–ø—Ä–æ–±—É–π—Ç–µ –µ—â–µ —Ä–∞–∑.")
        return handle_accept_invitation()

# –§—É–Ω–∫—Ü–∏—è. –í–≤–æ–¥ –Ω–µ—Å–∫–æ–ª—å–∫–∏—Ö –æ–ø—Ä–∞–≤–¥–∞–Ω–∏–π
def handle_propose():
    print("\n üíØ –û –Ω–µ—Ç! –û–Ω–∞ –ø–æ–¥–æ–∑—Ä–µ–≤–∞–µ—Ç –≤–∞—Å –≤ –Ω–µ–≤–µ—Ä–Ω–æ—Å—Ç–∏, –≤–∞–º –Ω—É–∂–Ω–æ –æ–ø—Ä–∞–≤–¥–∞—Ç—å—Å—è, –≤–≤–µ–¥–∏—Ç–µ 3 –ø—Ä–∏—á–∏–Ω—ã, –ø–æ—á–µ–º—É –≤—ã —Ö–æ–¥–∏–ª–∏–ª –≤ —Ä–µ—Å—Ç–æ—Ä–∞–Ω —Å –¥—Ä—É–≥–æ–π –¥–µ–≤—É—à–∫–æ–πü´£:")
    # –°–æ–∑–¥–∞–Ω–∏–µ –Ω–æ–≤–æ–≥–æ —Å–ª–æ–≤–∞—Ä—è
    excuses = {}
    # –í–≤–æ–¥ —Ç—Ä–µ—Ö –æ–ø—Ä–∞–≤–¥–∞–Ω–∏–π –∏ –ø–æ–º–µ—â–µ–Ω–∏–µ –∏—Ö –≤ —Å–ª–æ–≤–∞—Ä—å
    excuses[1] = input("–í–≤–µ–¥–∏—Ç–µ –ø–µ—Ä–≤–æ–µ –æ–ø—Ä–∞–≤–¥–∞–Ω–∏–µüíÉüèª: ")
    excuses[2] = input("–í–≤–µ–¥–∏—Ç–µ –≤—Ç–æ—Ä–æ–µ –æ–ø—Ä–∞–≤–¥–∞–Ω–∏–µü¶πüèª: ")
    excuses[3] = input("–í–≤–µ–¥–∏—Ç–µ —Ç—Ä–µ—Ç—å–µ –æ–ø—Ä–∞–≤–¥–∞–Ω–∏–µüëç: ")

    # –æ—Ç–∫—Ä—ã—Ç–∏–µ —Ñ–∞–ª–∞, —Å–æ–∑–¥–∞–Ω–∏–µ –µ—Å–ª–∏ –Ω–µ —Å—É—â–µ—Å—Ç–≤—É–µ—Ç –ò –∑–∞–ø–∏—Å—å –≤ —Ñ–∞–π–ª
    # –Ω–∞–∑–≤–∞–Ω–∏–µ —Ñ–∞–π–ª–∞ excuses.txt
    with open(name_file+".txt", "a") as file:
        file.write(f"\n2. –°–ø–∏—Å–æ–∫ –æ–ø—Ä–∞–≤–¥–∞–Ω–∏–π:")
        for i in range(1, 4):
            file.write(f"\n–û–ø—Ä–∞–≤–¥–∞–Ω–∏–µ {i}: {excuses[i]}\n")

    return "girlfriend_agrees"

# –§—É–Ω–∫—Ü–∏—è. –í—ã–±–∏—Ä–∞–µ–º —Å–ø–∏—Å–æ–∫ –±–ª—é–¥
def handle_decline_invitation():
    # —ç—Ç–æ –ø—É—Å—Ç–æ–π —Å–ø–∏—Å–æ–∫, –≤ –∫–æ—Ç–æ—Ä—ã–π –º–æ–∂–Ω–æ –¥–æ–±–∞–≤–ª—è—Ç—å —ç–ª–µ–º–µ–Ω—Ç—ã –ø–æ–∑–∂–µ
    dishes = []
    #  —ç—Ç–æ —Å–ø–∏—Å–æ–∫ —Å—Ç—Ä–æ–∫, –∫–∞–∂–¥–∞—è –∏–∑ –∫–æ—Ç–æ—Ä—ã—Ö –ø—Ä–µ–¥—Å—Ç–∞–≤–ª—è–µ—Ç –Ω–∞–∑–≤–∞–Ω–∏–µ –±–ª—é–¥–∞
    dish_names = ["–°–∞–ª–∞—Ç —Å –º–æ—Ä–µ–ø—Ä–æ–¥—É–∫—Ç–∞–º–∏", "–ö—Ä–µ–º-—Å—É–ø –∏–∑ —Ç—ã–∫–≤—ã", "–§–∏–ª–µ –º–∏–Ω—å–æ–Ω", "–†–∏–∑–æ—Ç—Ç–æ —Å –≥—Ä–∏–±–∞–º–∏ –∏ —Ç—Ä—é—Ñ–µ–ª—å–Ω—ã–º –º–∞—Å–ª–æ–º", "–¢–∏—Ä–∞–º–∏—Å—É"]
    print("\n üíØ –í—ã–±–µ—Ä–∏—Ç–µ –±–ª—é–¥–∞ –∏–∑ –ø—Ä–µ–¥–ª–æ–∂–µ–Ω–Ω—ã—Öüëç:")
    for i, dish in enumerate(dish_names, 1):
        print(f"{i}. {dish}")
    for _ in range(3):
        dish_choice = input("–í–≤–µ–¥–∏—Ç–µ –Ω–æ–º–µ—Ä –≤—ã–±—Ä–∞–Ω–Ω–æ–≥–æ –±–ª—é–¥–∞: ")
        if dish_choice in ["1", "2", "3", "4", "5"]:
            dishes.append(dish_names[int(dish_choice) - 1])
        else:
            print("–ù–µ–ø—Ä–∞–≤–∏–ª—å–Ω—ã–π –≤—ã–±–æ—Ä. –ü–æ–ø—Ä–æ–±—É–π—Ç–µ –µ—â–µ —Ä–∞–∑.")
            return handle_decline_invitation()

    # –æ—Ç–∫—Ä—ã—Ç–∏–µ —Ñ–∞–ª–∞, —Å–æ–∑–¥–∞–Ω–∏–µ –µ—Å–ª–∏ –Ω–µ —Å—É—â–µ—Å—Ç–≤—É–µ—Ç –ò –∑–∞–ø–∏—Å—å –≤ —Ñ–∞–π–ª
    # –Ω–∞–∑–≤–∞–Ω–∏–µ —Ñ–∞–π–ª–∞ selected_dishes.txt
    with open(name_file+".txt", "a") as file:
        for dish in dishes:
            file.write(f"\n3. –°–ø–∏—Å–æ–∫ –±–ª—é–¥:")
            file.write(f"\n{dish}\n")

    if dish_choice == "5":
        return "dessert_fight"
    else:
        return "fight_starts"

# –§—É–Ω–∫—Ü–∏—è –¥–ª—è –≤–≤–æ–¥–∞ –ø—Ä–µ–¥–º–µ—Ç–æ–≤ –∏–∑ –±–∏–±–ª–∏–æ—Ç–µ–∫–∏, —Å –º–µ—Ç–æ–¥–æ–º –º–Ω–æ–∂–µ—Å—Ç–≤
def input_library_items():
    library_items = ["–∫–ª—é—á", "—Ç–µ–ª–µ—Ñ–æ–Ω", "–∫–æ—à–µ–ª–µ–∫", "–ø–∞—Å–ø–æ—Ä—Ç"]
    user_items = input("–í–≤–µ–¥–∏—Ç–µ –ø—Ä–µ–¥–º–µ—Ç—ã –∏–∑ –±–∏–±–ª–∏–æ—Ç–µ–∫–∏ —á–µ—Ä–µ–∑ –∑–∞–ø—è—Ç—É—é: ").split(",")
    # strip() - –º–µ—Ç–æ–¥ –¥–ª—è —É–¥–∞–ª–µ–Ω–∏—è –ø—Ä–æ–±–µ–ª–æ–≤ –≤ –Ω–∞—á–∞–ª–µ –∏ –≤ –∫–æ–Ω—Ü–µ —Å—Ç—Ä–æ–∫–∏
    user_items = {item.strip() for item in user_items}  # –ü—Ä–µ–æ–±—Ä–∞–∑—É–µ–º –≤–≤–æ–¥ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è –≤ –º–Ω–æ–∂–µ—Å—Ç–≤–æ
    valid_items = user_items.intersection(library_items)  # –ü—Ä–æ–≤–µ—Ä–∫–∞ –ø–µ—Ä–µ—Å–µ—á–µ–Ω–∏—è —Å –º–Ω–æ–∂–µ—Å—Ç–≤–æ–º –¥–æ–ø—É—Å—Ç–∏–º—ã—Ö –ø—Ä–µ–¥–º–µ—Ç–æ–≤
    print(f"–ü—Ä–µ–¥–º–µ—Ç—ã, –∫–æ—Ç–æ—Ä—ã–µ –≤—ã –≤–∑—è–ª–∏: {', '.join(valid_items)}")
    # –ó–∞–ø–∏—Å—å –≤–∞–ª–∏–¥–Ω—ã—Ö –ø—Ä–µ–¥–º–µ—Ç–æ–≤ –≤ —Ñ–∞–π–ª
    with open(name_file+".txt", "a") as file:
        file.write(f"\n1. –ü—Ä–µ–¥–º–µ—Ç—ã –∏–∑ –±–∏–±–ª–∏–æ—Ç–µ–∫–∏:")
        file.write(f"\n{', '.join(valid_items)}\n")


# –§—É–Ω–∫—Ü–∏—è. –û—Å–Ω–æ–≤–Ω–∞—è —Ñ—É–Ω–∫—Ü–∏—è
def start_story():
    # –û—á–∏—â–∞–µ–º —Ñ–∞–π–ª –ø–µ—Ä–µ–¥ –Ω–∞—á–∞–ª–æ–º –Ω–æ–≤–æ–π –∏—Å—Ç–æ—Ä–∏–∏
    with open(name_file+".txt", "w") as file:
        pass
    current_scene = "start"
    final_scene = None
    while current_scene:
        # if –¥–ª—è –æ—Å–æ–±—ã—Ö —Å—Ü–µ–Ω –∫–æ–≥–¥–∞ –Ω—É–∂–Ω–æ —Å–¥–µ–ª–∞—Ç—å –∫–∞–∫–∏–µ-—Ç–æ –¥–µ–π—Å—Ç–≤–∏—è, –∞ –Ω–µ –ø—Ä–æ—Å—Ç–æ –≤—ã–±—Ä–∞—Ç—å —Å—Ü–µ–Ω–∞—Ä–∏–π
        if current_scene == "boat_dinner":
            current_scene = handle_boat_dinner()
        elif current_scene == "accept_invitation":
            current_scene = handle_accept_invitation()
        elif current_scene == "propose":
            current_scene = handle_propose()
        elif current_scene == "decline_invitation":
            current_scene = handle_decline_invitation()
        else:
            next_scene = display_scene(current_scene)
            if next_scene is None:
                final_scene = current_scene
                break
            else:
                current_scene = next_scene
    # –ï—Å–ª–∏ —ç—Ç–æ –∫–æ–Ω—Ü–æ–≤–∫–∞ —Ç–æ –º—ã –æ—Ç–æ–±—Ä–∞–∂–µ–º —á—Ç–æ —ç—Ç–æ –∫–æ–Ω–µ—Ü –∏ –∑–∞–ø–∏—Å—ã–≤–∞–µ–º –≤ —Ñ–∞–π–ª
    if final_scene:
        print("–ò—Å—Ç–æ—Ä–∏—è –∑–∞–≤–µ—Ä—à–µ–Ω–∞. –°–ø–∞—Å–∏–±–æ –∑–∞ –∏–≥—Ä—É! üòä")
        # –æ—Ç–∫—Ä—ã—Ç–∏–µ —Ñ–∞–ª–∞, —Å–æ–∑–¥–∞–Ω–∏–µ –µ—Å–ª–∏ –Ω–µ —Å—É—â–µ—Å—Ç–≤—É–µ—Ç –ò –∑–∞–ø–∏—Å—å –≤ —Ñ–∞–π–ª
        # –Ω–∞–∑–≤–∞–Ω–∏–µ —Ñ–∞–π–ª–∞ final_scene.txt
        with open(name_file+".txt", "a") as file:
            file.write(f"\n4. –ö–æ–Ω–µ—á–Ω–∞—è —Å—Ü–µ–Ω–∞:\n{stories[final_scene]['title']}\n{stories[final_scene]['description']}\n")

# –∑–∞–ø—É—Å–∫ —Ñ—É–Ω–∫—Ü–∏–∏ –¥–ª—è —Ç–æ–≥–æ —á—Ç–æ–±—ã –∑–∞–ø—É—Å—Ç–∏—Ç—å –∫–æ–¥
start_story()
```
