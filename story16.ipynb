```
#### Использовать любые библиотеки / импортированные функции запрещено.
#### => СОБЛЮДЕНО Импорты указываются в начале файла (import ...), их тут нету поэтому условие соблюдено 

#### Для проверки кода предоставлен файл в формате .ipynb
#### => СОБЛЮДЕНО название файла в формате .ipynb 

#### В итоге работы представлен цельный сюжет с минимум (3) тремя различными концовками (на достижение которых может влиять пользователь своим выбором).
#### => СОБЛЮДЕНО концовки можно найти по префиксу (КОНЦОВКА:) - их 6 штук

#### В коде использован хотя бы один цикл while. Например, сюжет
# такой: грабитель пытается подобрать пароль к сейфу, пока не
# закончится время. Если время вышло, то его могут поймать
# жильцы, а если грабитель успел вовремя, то он получает
# драгоценности из сейфа.
#### => СОБЛЮДЕНО их 2 штуки

#### В коде использован хотя бы один цикл for. Например, сюжет
# такой: на этаже есть 4 двери и грабитель проверяет каждую на
# предмет наличия за дверью того, что можно украсть. Он
# может даже зайти в каждую комнату и осмотреться.
#### => СОБЛЮДЕНО 8 штук

#### В коде использовано хотя бы 
# два метода списков / 
# два методамножеств / 
# один метод списков и один метод множеств.
# Например, списком в данном случае может быть сумка
# грабителя, в которую он складывает украденное.
#### => СОБЛЮДЕНО два метода списков =  append 2 штуки (не было сказано два РАЗНЫХ метода)
#### => СОБЛЮДЕНО один метод множеств intersection

#### В коде создается хотя бы один новый словарь. Например, если
# грабителя поймала полиция, то она может записать в словарь
# названия статей, которые он нарушил, и наказание по каждой
# статье.
#### => СОБЛЮДЕНО название словаря excuses, в handle_propose, в создание оправданий

### В коде использовано хотя бы 4 условных конструкций.
# Например, сюжет такой: грабитель убегает от преследования и
# выбирает, свернуть направо или налево — если он свернет
# направо, то его задержит полиция, а если он свернет направо,
# то встретит мафию, которая заберет у него все награбленное.
# Таких разветвлений в сюжете может быть и больше.
#### => СОБЛЮДЕНО , условные конструкции это if / else / elif 

#### В коде использован хотя бы один готовый словарь. Например,
# в виде словаря может быть представлена адресная книга, из
# которой грабитель выбирает себе потенциальную жертву.
#### => СОБЛЮДЕНО словарь называкется stories

#### Итоговый сюжет (со всеми выборами пользователя) и
# получившаяся концовка автоматически записаны в файл.
#### => СОБЛЮДЕНО записывается кнцовка в один файл а другие данные в другие файлы, чтоыб точно записалось. Всего 4 файла разных




# Сама история со сценариями, как словаря в виде графа (дерева), 
# так как графа (дерево) позволяет перебирать и выбирать разные варианты, 
# также это позволяет не деоать много if под каждый сценарий
stories = {
    "start": {
        "title": "Мы в отеле с партнером",
        "description": "Люси и Марк приезжают в роскошный отель «Белый Лотос» на побережье, они не женаты. Надо взять что-то с собой выходя из номера. Что вы положите в сумку?\nВы выбираете, как хотели бы провести вечер",
        "choices": {
            "1": "restaurant_dinner",
            "2": "boat_dinner"
        }
    },
    "restaurant_dinner": {
        "title": "Отправиться на ужин в ресторан отеля",
        "description": "Люси и Марк наслаждаются роскошной едой и встречают богатого бизнесмена, который предлагает им экскурсию по окрестностям. Что вы выберете?",
        "choices": {
            "1": "accept_invitation",
            "2": "decline_invitation"
        }
    },
    "boat_dinner": {
        "title": "Устроить ужин на лодке",
        "description": "Вы пришли на борт, администратор просит вас ввести данные с карты, вы пытаетесь вспомнить пароль.",
        "choices": {
            "1": "remember_code",
            "2": "forget_code"
        }
    },
    "accept_invitation": {
        "title": "Принять приглашение",
        "description": "Вы что-то обсуждаете с бизнесменом. У вас есть друг который знает необходимую информацию. Вы выбираете из контактов кому позвонить. Выберите из следующих имен: Мартина, Луиза, Винни, Майк.",
        "choices": {
            "1": "call_martin",
            "2": "call_louise",
            "3": "call_winnie",
            "4": "call_mike"
        }
    },
    "decline_invitation": {
        "title": "Отказаться и остаться на ужине",
        "description": "Кажется, пришло время выбирать блюда. Что вы предпочете?\n\n1. Салат с морепродуктами\n2. Крем-суп из тыквы\n3. Филе миньон\n4. Ризотто с грибами и трюфельным маслом\n5. Тирамису",
        "choices": {
            "1": "fight_starts",
            "2": "dessert_fight"
        }
    },
    "dessert_fight": {
        "title": "Начало скандала",
        "description": "Один из постояльцев устраивает драку из-за десерта, вы не отвечаете и вас арестовывают.",
        "choices": {}
    },
    "fight_starts": {
        "title": "Начало драки",
        "description": "Один из постояльцев устраивает драку, но вы вмешиваетесь и на вас не обращают внимание.",
        "choices": {}
    },
    "remember_code": {
        "title": "Вспомнил код",
        "description": "Администратор пропускает вас за столик. Наслаждаясь видом, вы думаете, принять решение которое так давно планировали.",
        "choices": {
            "1": "propose",
            "2": "not_propose"
        }
    },
    "forget_code": {
        "title": "Не вспомнил код",
        "description": "Администратор не пропускает вас, и вас бросает девушка. Вы проводите отпуск один.",
        "choices": {}
    },
    "propose": {
        "title": "Сделать предложение",
        "description": "О нет! Она подозревает вас в неверности, вам нужно оправдаться. Введите 3 причины, почему вы ходили в ресторан с другой девушкой.",
        "choices": {
            "1": "give_excuses"
        }
    },
    "not_propose": {
        "title": "Не делать предложение",
        "description": "Она так долго ждала вашего предложения, увы, это был последний шанс.",
        "choices": {}
    },
    "give_excuses": {
        "title": "Дать оправдания",
        "description": "Она требует оправдания для каждой из трех измен. Придумайте три оправдания.",
        "choices": {
            "1": "excuse_1",
            "2": "excuse_2",
            "3": "excuse_3"
        }
    },
    "excuse_1": {
        "title": "Первое оправдание",
        "description": "Введите первое оправдание.",
        "choices": {}
    },
    "excuse_2": {
        "title": "Второе оправдание",
        "description": "Введите второе оправдание.",
        "choices": {}
    },
    "excuse_3": {
        "title": "Третье оправдание",
        "description": "Введите третье оправдание.",
        "choices": {
            "1": "girlfriend_agrees"
        }
    },
    "girlfriend_agrees": {
        "title": "Она соглашается",
        "description": "Фух! Она накрутила себя на нервах... Конечно согласна.",
        "choices": {}
    },
    "call_martin": {
        "title": "Позвонить Мартину",
        "description": "Вы позвонили вашему другу, заклятому врагу бизнесмена, и он вас хочет убить.",
        "choices": {}
    },
    "call_louise": {
        "title": "Позвонить Луизе",
        "description": "Вы звоните верному другу и получаете доверие бизнесмена.",
        "choices": {}
    },
    "call_winnie": {
        "title": "Позвонить Винни",
        "description": "Вы позвонили вашему другу, заклятому врагу бизнесмена, и он вас хочет убить.",
        "choices": {}
    },
    "call_mike": {
        "title": "Позвонить Майку",
        "description": "Вы звоните верному другу и получаете доверие бизнесмена.",
        "choices": {}
    },
    "final_arrest": {
        "title": "Вас арестовывают",
        "description": "КОНЦОВКА: Вас арестовали за участие в драке. Это конец вашей истории.",
        "choices": {}
    },
    "final_ignored": {
        "title": "Игнорирование",
        "description": "КОНЦОВКА: Вас не заметили и вы провели вечер в покое. Это конец вашей истории.",
        "choices": {}
    },
    "final_alone": {
        "title": "Оставаться одному",
        "description": "КОНЦОВКА: Вас бросает девушка, и вы проводите отпуск один. Это конец вашей истории.",
        "choices": {}
    },
    "final_no_proposal": {
        "title": "Нет предложения",
        "description": "КОНЦОВКА: Она так долго ждала вашего предложения, увы, это был последний шанс. Это конец вашей истории.",
        "choices": {}
    },
    "final_attack": {
        "title": "Атака",
        "description": "КОНЦОВКА: Ваш заклятый враг атакует вас, и вы вынуждены бороться за свою жизнь. Это конец вашей истории.",
        "choices": {}
    },
    "final_apartment": {
        "title": "Получение квартиры",
        "description": "КОНЦОВКА: Вы получаете доверие бизнесмена, и он дарит вам квартиру. Это конец вашей истории.",
        "choices": {}
    }
}

contacts = ["Мартин", "Луиза", "Винни", "Майк"]  # Список контактов

correct_code = "1205"  # Код, который нужно угадать

name_file = "story_file"

# input() - метод ввода / с клавиатуры текст получить
# print() - метод вывода / отбражения текста 

# Функция. Отображаем описание сценария и выбор действий
def display_scene(scene):
    print(f"\n 💯 {stories[scene]['title']}\n{stories[scene]['description']}\n")
    if scene == "start":
        input_library_items()  # Ввод предметов из библиотеки
        print(f"\nВыберите следующее действие\n")
    choices = list(stories[scene]['choices'].keys())
    if len(choices) == 1:
        # Если есть только один выбор то автоматически выбирается он
        next_scene = stories[scene]['choices'][choices[0]]
        print(f"Единственный выбор: {stories[next_scene]['title']}")
        return next_scene
    elif choices:
        # Отображаем все варианты которые есть
        for choice in choices:
            print(f"{choice}. {stories[stories[scene]['choices'][choice]]['title']}")
        user_choice = input("Выберите вариант: ")
        if user_choice in choices:
            return stories[scene]['choices'][user_choice]
        else:
            print("Неправильный выбор🙅🏻‍♂️. Попробуйте еще раз.")
            return display_scene(scene)
    else:
        return None

# Функция. Ввод кода
def handle_boat_dinner():
    attempt = ""
    number_of_attempts = 3  # Ограничение на количество попыток
    print("\n 💯 Он оплачивает картой. Забыл пароль 🤦🏻‍♀️")
    while attempt != correct_code and number_of_attempts > 0:
        attempt = input(f"подбирает код (осталось попыток {number_of_attempts}): ")
        if attempt == correct_code:
            print("Вы вспомнили код и прошли на ужин.")
            return "remember_code"
        else:
            # Убераем минус одну попытку
            number_of_attempts -= 1
            if number_of_attempts == 0:
                print("Неправильный код. Вы исчерпали все попытки.")
                return "forget_code"
            else:
                # Обрабатываем случай если ввел неправильно выбор
                print("Неправильный код. Попробуйте еще раз.")

# Выбор кому позвонить
def handle_accept_invitation():
    print("\n 💯 Вы что-то обсуждаете с бизнесменом. У вас есть друг который знает необходимую информацию. Вы выбираете из контактов кому позвонить. Выбирете из следующих имен:\n")
    print("Выбираете, кому позвонить из списка контактов📱:")
    for i, contact in enumerate(contacts, 1):
        print(f"{i}. {contact}")
    user_choice = input("Введите номер контакта👨🏽‍💻: ")
    if user_choice == "1":
        return "call_martin"
    elif user_choice == "2":
        return "call_louise"
    elif user_choice == "3":
        return "call_winnie"
    elif user_choice == "4":
        return "call_mike"
    else:
        # Обрабатываем случай если ввел неправильно выбор
        print("Неправильный выбор. Попробуйте еще раз.")
        return handle_accept_invitation()

# Функция. Ввод нескольких оправданий
def handle_propose():
    print("\n 💯 О нет! Она подозревает вас в неверности, вам нужно оправдаться, введите 3 причины, почему вы ходилил в ресторан с другой девушкой🫣:")
    # Создание нового словаря
    excuses = {}
    # Ввод трех оправданий и помещение их в словарь
    excuses[1] = input("Введите первое оправдание💃🏻: ")
    excuses[2] = input("Введите второе оправдание🦹🏻: ")
    excuses[3] = input("Введите третье оправдание👍: ")

    # открытие фала, создание если не существует И запись в файл
    # название файла excuses.txt
    with open(name_file+".txt", "a") as file:
        file.write(f"\n2. Список оправданий:")
        for i in range(1, 4):
            file.write(f"\nОправдание {i}: {excuses[i]}\n")

    return "girlfriend_agrees"

# Функция. Выбираем список блюд
def handle_decline_invitation():
    # это пустой список, в который можно добавлять элементы позже
    dishes = []
    #  это список строк, каждая из которых представляет название блюда
    dish_names = ["Салат с морепродуктами", "Крем-суп из тыквы", "Филе миньон", "Ризотто с грибами и трюфельным маслом", "Тирамису"]
    print("\n 💯 Выберите блюда из предложенных👍:")
    for i, dish in enumerate(dish_names, 1):
        print(f"{i}. {dish}")
    for _ in range(3):
        dish_choice = input("Введите номер выбранного блюда: ")
        if dish_choice in ["1", "2", "3", "4", "5"]:
            dishes.append(dish_names[int(dish_choice) - 1])
        else:
            print("Неправильный выбор. Попробуйте еще раз.")
            return handle_decline_invitation()

    # открытие фала, создание если не существует И запись в файл
    # название файла selected_dishes.txt
    with open(name_file+".txt", "a") as file:
        for dish in dishes:
            file.write(f"\n3. Список блюд:")
            file.write(f"\n{dish}\n")

    if dish_choice == "5":
        return "dessert_fight"
    else:
        return "fight_starts"

# Функция для ввода предметов из библиотеки, с методом множеств
def input_library_items():
    library_items = ["ключ", "телефон", "кошелек", "паспорт"]
    user_items = input("Введите предметы из библиотеки через запятую: ").split(",")
    # strip() - метод для удаления пробелов в начале и в конце строки
    user_items = {item.strip() for item in user_items}  # Преобразуем ввод пользователя в множество
    valid_items = user_items.intersection(library_items)  # Проверка пересечения с множеством допустимых предметов
    print(f"Предметы, которые вы взяли: {', '.join(valid_items)}")
    # Запись валидных предметов в файл
    with open(name_file+".txt", "a") as file:
        file.write(f"\n1. Предметы из библиотеки:")
        file.write(f"\n{', '.join(valid_items)}\n")


# Функция. Основная функция
def start_story():
    # Очищаем файл перед началом новой истории
    with open(name_file+".txt", "w") as file:
        pass
    current_scene = "start"
    final_scene = None
    while current_scene:
        # if для особых сцен когда нужно сделать какие-то действия, а не просто выбрать сценарий
        if current_scene == "boat_dinner":
            current_scene = handle_boat_dinner()
        elif current_scene == "accept_invitation":
            current_scene = handle_accept_invitation()
        elif current_scene == "propose":
            current_scene = handle_propose()
        elif current_scene == "decline_invitation":
            current_scene = handle_decline_invitation()
        else:
            next_scene = display_scene(current_scene)
            if next_scene is None:
                final_scene = current_scene
                break
            else:
                current_scene = next_scene
    # Если это концовка то мы отображем что это конец и записываем в файл
    if final_scene:
        print("История завершена. Спасибо за игру! 😊")
        # открытие фала, создание если не существует И запись в файл
        # название файла final_scene.txt
        with open(name_file+".txt", "a") as file:
            file.write(f"\n4. Конечная сцена:\n{stories[final_scene]['title']}\n{stories[final_scene]['description']}\n")

# запуск функции для того чтобы запустить код
start_story()
```
