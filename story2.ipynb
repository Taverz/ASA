```
# Определение истории как словаря
stories = {
    "start": {
        "title": "Мы в отеле с партнером",
        "description": "Люси и Марк приезжают в роскошный отель «Белый Лотос» на побережье, они не женаты. Надо взять что-то с собой выходя из номера. Что вы положите в сумку?",
        "choices": {
            "1": "restaurant_dinner",
            "2": "boat_dinner"
        }
    },
    "restaurant_dinner": {
        "title": "Отправиться на ужин в ресторан отеля",
        "description": "Люси и Марк наслаждаются роскошной едой и встречают богатого бизнесмена, который предлагает им экскурсию по окрестностям. Что вы выберете?",
        "choices": {
            "1": "accept_invitation",
            "2": "decline_invitation"
        }
    },
    "boat_dinner": {
        "title": "Устроить ужин на лодке",
        "description": "Вы пришли на борт, администратор просит вас ввести данные с карты, вы пытаетесь вспомнить пароль.",
        "choices": {
            "1": "remember_code",
            "2": "forget_code"
        }
    },
    "accept_invitation": {
        "title": "Принять приглашение",
        "description": "Вы что-то обсуждаете с бизнесменом. У вас есть друг который знает необходимую информацию. Вы выбираете из контактов кому позвонить. Выберите из следующих имен: Мартина, Луиза, Винни, Майк.",
        "choices": {}
    },
    "decline_invitation": {
        "title": "Отказаться и остаться на ужине",
        "description": "Один из постояльцев устраивает драку, вы не отвечаете и вас арестовывают.",
        "choices": {
            "1": "final_arrest"
        }
    },
    "remember_code": {
        "title": "Вспомнить код",
        "description": "Администратор пропускает вас за столик. Наслаждаясь видом, вы думаете, принять решение которое так давно планировали.",
        "choices": {
            "1": "propose",
            "2": "not_propose"
        }
    },
    "forget_code": {
        "title": "Не вспомнить код",
        "description": "Администратор не пропускает вас, и вас бросает девушка. Вы проводите отпуск один.",
        "choices": {
            "1": "final_alone"
        }
    },
    "propose": {
        "title": "Сделать предложение",
        "description": "О нет! Она подозревает вас в неверности, вам нужно оправдаться. Введите 3 причины, почему вы ходили в ресторан с другой девушкой.",
        "choices": {
            "1": "give_excuses"
        }
    },
    "not_propose": {
        "title": "Не делать предложение",
        "description": "Она так долго ждала вашего предложения, увы, это был последний шанс.",
        "choices": {
            "1": "final_no_proposal"
        }
    },
    "give_excuses": {
        "title": "Дать оправдания",
        "description": "Она требует оправдания для каждой из трех измен. Придумайте три оправдания.",
        "choices": {
            "1": "excuse_1",
            "2": "excuse_2",
            "3": "excuse_3"
        }
    },
    "excuse_1": {
        "title": "Первое оправдание",
        "description": "Введите первое оправдание.",
        "choices": {}
    },
    "excuse_2": {
        "title": "Второе оправдание",
        "description": "Введите второе оправдание.",
        "choices": {}
    },
    "excuse_3": {
        "title": "Третье оправдание",
        "description": "Введите третье оправдание.",
        "choices": {
            "1": "girlfriend_agrees"
        }
    },
    "girlfriend_agrees": {
        "title": "Она соглашается",
        "description": "Фух! Она накрутила себя на нервах... Конечно согласна.",
        "choices": {
            "1": "final_agreement"
        }
    },
    "call_martin": {
        "title": "Позвонить Мартину",
        "description": "Вы позвонили вашему другу, заклятому врагу бизнесмена, и он вас хочет убить.",
        "choices": {
            "1": "final_attack"
        }
    },
    "call_louise": {
        "title": "Позвонить Луизе",
        "description": "Вы звоните верному другу и получаете доверие бизнесмена. Он дарит вам квартиру.",
        "choices": {
            "1": "final_apartment"
        }
    },
    "call_winnie": {
        "title": "Позвонить Винни",
        "description": "Вы позвонили вашему другу, заклятому врагу бизнесмена, и он вас хочет убить.",
        "choices": {
            "1": "final_attack"
        }
    },
    "call_mike": {
        "title": "Позвонить Майку",
        "description": "Вы звоните верному другу и получаете доверие бизнесмена. Он дарит вам квартиру.",
        "choices": {
            "1": "final_apartment"
        }
    },
    "final_arrest": {
        "title": "Вас арестовывают",
        "description": "КОНЦОВКА номер 1: Вас арестовали за участие в драке. Это конец вашей истории.",
        "choices": {}
    },
    "final_alone": {
        "title": "Оставаться одному",
        "description": "КОНЦОВКА номер 2: Вас бросает девушка, и вы проводите отпуск один. Это конец вашей истории.",
        "choices": {}
    },
    "final_no_proposal": {
        "title": "Нет предложения",
        "description": "КОНЦОВКА номер 3: Она так долго ждала вашего предложения, увы, это был последний шанс. Это конец вашей истории.",
        "choices": {}
    },
    "final_agreement": {
        "title": "Согласие",
        "description": "КОНЦОВКА номер 4: Она согласилась с предложением и вы счастливы вместе. Это конец вашей истории.",
        "choices": {}
    },
    "final_attack": {
        "title": "Атака",
        "description": "КОНЦОВКА номер 5: Ваш заклятый враг атакует вас, и вы вынуждены бороться за свою жизнь. Это конец вашей истории.",
        "choices": {}
    },
    "final_apartment": {
        "title": "Получение квартиры",
        "description": "КОНЦОВКА номер 6: Вы получаете доверие бизнесмена, и он дарит вам квартиру. Это конец вашей истории.",
        "choices": {}
    }
}


contacts = ["Мартин", "Луиза", "Винни", "Майк"]  # Список контактов

# Функция для отображения текущей сцены и получения выбора пользователя
def display_scene(scene):
    print(f"\nОписание👨🏽‍💻:\n {stories[scene]['description']}")
    if scene == "boat_dinner":
        return handle_boat_dinner()
    elif scene == "accept_invitation":
        return handle_accept_invitation()
    elif scene == "propose":
        return handle_propose()

    choices = stories[scene]['choices']
    if len(choices) == 1:
        next_scene = list(choices.values())[0]
        print(f"Автоматический выбор: {stories[next_scene]['title']}")
        return next_scene
    elif choices:
        for choice, next_scene in choices.items():
            print(f"{choice}. {stories[next_scene]['title']}")
        user_choice = input("Выберите вариант: ")
        if user_choice in choices:
            return choices[user_choice]
        else:
            print("Неправильный выбор. Попробуйте еще раз.")
            return display_scene(scene)
    else:
        return None

# Функция для обработки сцены "Устроить ужин на лодке"
def handle_boat_dinner():
    correct_code = "1205"  # Код, который нужно угадать
    attempt = ""
    while attempt != correct_code:
        attempt = input("Он оплачивает картой. Забыл пароль, подбирает код: ")
        if attempt == correct_code:
            print("Вы вспомнили код и прошли на ужин.")
        else:
            print("Неправильный код. Попробуйте еще раз.")

# Функция для обработки сцены "Принять приглашение"
def handle_accept_invitation():
    print("Выбираете, кому позвонить из списка контактов:")
    for i, contact in enumerate(contacts, 1):
        print(f"{i}. {contact}")
    user_choice = input("Введите номер контакта👨🏽‍💻: ")
    if user_choice == "1":
        return "call_martin"
    elif user_choice == "2":
        return "call_louise"
    elif user_choice == "3":
        return "call_winnie"
    elif user_choice == "4":
        return "call_mike"
    else:
        print("Неправильный выбор. Попробуйте еще раз.")
        return

# Функция для обработки сцены "Сделать предложение"
def handle_propose():
    print("О нет! Она подозревает вас в неверности, вам нужно оправдаться. Введите 3 причины, почему вы ходили в ресторан с другой девушкой.")
    excuses = {}
    excuses[1] = input("Введите первое оправдание👍: ")
    excuses[2] = input("Введите второе оправдание👍: ")
    excuses[3] = input("Введите третье оправдание👍: ")

    # Запись оправданий в файл
    with open("excuses.txt", "w") as file:
        for i in range(1, 4):
            file.write(f"Оправдание {i}: {excuses[i]}\n")

    return "girlfriend_agrees"

# Основная функция
def start_story():
    current_scene = "start"
    while current_scene:
        current_scene = display_scene(current_scene)

# Запуск истории
start_story()
```
