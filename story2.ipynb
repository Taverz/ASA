```
# –û–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∏—Å—Ç–æ—Ä–∏–∏ –∫–∞–∫ —Å–ª–æ–≤–∞—Ä—è
stories = {
    "start": {
        "title": "–ú—ã –≤ –æ—Ç–µ–ª–µ —Å –ø–∞—Ä—Ç–Ω–µ—Ä–æ–º",
        "description": "–õ—é—Å–∏ –∏ –ú–∞—Ä–∫ –ø—Ä–∏–µ–∑–∂–∞—é—Ç –≤ —Ä–æ—Å–∫–æ—à–Ω—ã–π –æ—Ç–µ–ª—å ¬´–ë–µ–ª—ã–π –õ–æ—Ç–æ—Å¬ª –Ω–∞ –ø–æ–±–µ—Ä–µ–∂—å–µ, –æ–Ω–∏ –Ω–µ –∂–µ–Ω–∞—Ç—ã. –ù–∞–¥–æ –≤–∑—è—Ç—å —á—Ç–æ-—Ç–æ —Å —Å–æ–±–æ–π –≤—ã—Ö–æ–¥—è –∏–∑ –Ω–æ–º–µ—Ä–∞. –ß—Ç–æ –≤—ã –ø–æ–ª–æ–∂–∏—Ç–µ –≤ —Å—É–º–∫—É?",
        "choices": {
            "1": "restaurant_dinner",
            "2": "boat_dinner"
        }
    },
    "restaurant_dinner": {
        "title": "–û—Ç–ø—Ä–∞–≤–∏—Ç—å—Å—è –Ω–∞ —É–∂–∏–Ω –≤ —Ä–µ—Å—Ç–æ—Ä–∞–Ω –æ—Ç–µ–ª—è",
        "description": "–õ—é—Å–∏ –∏ –ú–∞—Ä–∫ –Ω–∞—Å–ª–∞–∂–¥–∞—é—Ç—Å—è —Ä–æ—Å–∫–æ—à–Ω–æ–π –µ–¥–æ–π –∏ –≤—Å—Ç—Ä–µ—á–∞—é—Ç –±–æ–≥–∞—Ç–æ–≥–æ –±–∏–∑–Ω–µ—Å–º–µ–Ω–∞, –∫–æ—Ç–æ—Ä—ã–π –ø—Ä–µ–¥–ª–∞–≥–∞–µ—Ç –∏–º —ç–∫—Å–∫—É—Ä—Å–∏—é –ø–æ –æ–∫—Ä–µ—Å—Ç–Ω–æ—Å—Ç—è–º. –ß—Ç–æ –≤—ã –≤—ã–±–µ—Ä–µ—Ç–µ?",
        "choices": {
            "1": "accept_invitation",
            "2": "decline_invitation"
        }
    },
    "boat_dinner": {
        "title": "–£—Å—Ç—Ä–æ–∏—Ç—å —É–∂–∏–Ω –Ω–∞ –ª–æ–¥–∫–µ",
        "description": "–í—ã –ø—Ä–∏—à–ª–∏ –Ω–∞ –±–æ—Ä—Ç, –∞–¥–º–∏–Ω–∏—Å—Ç—Ä–∞—Ç–æ—Ä –ø—Ä–æ—Å–∏—Ç –≤–∞—Å –≤–≤–µ—Å—Ç–∏ –¥–∞–Ω–Ω—ã–µ —Å –∫–∞—Ä—Ç—ã, –≤—ã –ø—ã—Ç–∞–µ—Ç–µ—Å—å –≤—Å–ø–æ–º–Ω–∏—Ç—å –ø–∞—Ä–æ–ª—å.",
        "choices": {
            "1": "remember_code",
            "2": "forget_code"
        }
    },
    "accept_invitation": {
        "title": "–ü—Ä–∏–Ω—è—Ç—å –ø—Ä–∏–≥–ª–∞—à–µ–Ω–∏–µ",
        "description": "–í—ã —á—Ç–æ-—Ç–æ –æ–±—Å—É–∂–¥–∞–µ—Ç–µ —Å –±–∏–∑–Ω–µ—Å–º–µ–Ω–æ–º. –£ –≤–∞—Å –µ—Å—Ç—å –¥—Ä—É–≥ –∫–æ—Ç–æ—Ä—ã–π –∑–Ω–∞–µ—Ç –Ω–µ–æ–±—Ö–æ–¥–∏–º—É—é –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é. –í—ã –≤—ã–±–∏—Ä–∞–µ—Ç–µ –∏–∑ –∫–æ–Ω—Ç–∞–∫—Ç–æ–≤ –∫–æ–º—É –ø–æ–∑–≤–æ–Ω–∏—Ç—å. –í—ã–±–µ—Ä–∏—Ç–µ –∏–∑ —Å–ª–µ–¥—É—é—â–∏—Ö –∏–º–µ–Ω: –ú–∞—Ä—Ç–∏–Ω–∞, –õ—É–∏–∑–∞, –í–∏–Ω–Ω–∏, –ú–∞–π–∫.",
        "choices": {}
    },
    "decline_invitation": {
        "title": "–û—Ç–∫–∞–∑–∞—Ç—å—Å—è –∏ –æ—Å—Ç–∞—Ç—å—Å—è –Ω–∞ —É–∂–∏–Ω–µ",
        "description": "–û–¥–∏–Ω –∏–∑ –ø–æ—Å—Ç–æ—è–ª—å—Ü–µ–≤ —É—Å—Ç—Ä–∞–∏–≤–∞–µ—Ç –¥—Ä–∞–∫—É, –≤—ã –Ω–µ –æ—Ç–≤–µ—á–∞–µ—Ç–µ –∏ –≤–∞—Å –∞—Ä–µ—Å—Ç–æ–≤—ã–≤–∞—é—Ç.",
        "choices": {
            "1": "final_arrest"
        }
    },
    "remember_code": {
        "title": "–í—Å–ø–æ–º–Ω–∏—Ç—å –∫–æ–¥",
        "description": "–ê–¥–º–∏–Ω–∏—Å—Ç—Ä–∞—Ç–æ—Ä –ø—Ä–æ–ø—É—Å–∫–∞–µ—Ç –≤–∞—Å –∑–∞ —Å—Ç–æ–ª–∏–∫. –ù–∞—Å–ª–∞–∂–¥–∞—è—Å—å –≤–∏–¥–æ–º, –≤—ã –¥—É–º–∞–µ—Ç–µ, –ø—Ä–∏–Ω—è—Ç—å —Ä–µ—à–µ–Ω–∏–µ –∫–æ—Ç–æ—Ä–æ–µ —Ç–∞–∫ –¥–∞–≤–Ω–æ –ø–ª–∞–Ω–∏—Ä–æ–≤–∞–ª–∏.",
        "choices": {
            "1": "propose",
            "2": "not_propose"
        }
    },
    "forget_code": {
        "title": "–ù–µ –≤—Å–ø–æ–º–Ω–∏—Ç—å –∫–æ–¥",
        "description": "–ê–¥–º–∏–Ω–∏—Å—Ç—Ä–∞—Ç–æ—Ä –Ω–µ –ø—Ä–æ–ø—É—Å–∫–∞–µ—Ç –≤–∞—Å, –∏ –≤–∞—Å –±—Ä–æ—Å–∞–µ—Ç –¥–µ–≤—É—à–∫–∞. –í—ã –ø—Ä–æ–≤–æ–¥–∏—Ç–µ –æ—Ç–ø—É—Å–∫ –æ–¥–∏–Ω.",
        "choices": {
            "1": "final_alone"
        }
    },
    "propose": {
        "title": "–°–¥–µ–ª–∞—Ç—å –ø—Ä–µ–¥–ª–æ–∂–µ–Ω–∏–µ",
        "description": "–û –Ω–µ—Ç! –û–Ω–∞ –ø–æ–¥–æ–∑—Ä–µ–≤–∞–µ—Ç –≤–∞—Å –≤ –Ω–µ–≤–µ—Ä–Ω–æ—Å—Ç–∏, –≤–∞–º –Ω—É–∂–Ω–æ –æ–ø—Ä–∞–≤–¥–∞—Ç—å—Å—è. –í–≤–µ–¥–∏—Ç–µ 3 –ø—Ä–∏—á–∏–Ω—ã, –ø–æ—á–µ–º—É –≤—ã —Ö–æ–¥–∏–ª–∏ –≤ —Ä–µ—Å—Ç–æ—Ä–∞–Ω —Å –¥—Ä—É–≥–æ–π –¥–µ–≤—É—à–∫–æ–π.",
        "choices": {
            "1": "give_excuses"
        }
    },
    "not_propose": {
        "title": "–ù–µ –¥–µ–ª–∞—Ç—å –ø—Ä–µ–¥–ª–æ–∂–µ–Ω–∏–µ",
        "description": "–û–Ω–∞ —Ç–∞–∫ –¥–æ–ª–≥–æ –∂–¥–∞–ª–∞ –≤–∞—à–µ–≥–æ –ø—Ä–µ–¥–ª–æ–∂–µ–Ω–∏—è, —É–≤—ã, —ç—Ç–æ –±—ã–ª –ø–æ—Å–ª–µ–¥–Ω–∏–π —à–∞–Ω—Å.",
        "choices": {
            "1": "final_no_proposal"
        }
    },
    "give_excuses": {
        "title": "–î–∞—Ç—å –æ–ø—Ä–∞–≤–¥–∞–Ω–∏—è",
        "description": "–û–Ω–∞ —Ç—Ä–µ–±—É–µ—Ç –æ–ø—Ä–∞–≤–¥–∞–Ω–∏—è –¥–ª—è –∫–∞–∂–¥–æ–π –∏–∑ —Ç—Ä–µ—Ö –∏–∑–º–µ–Ω. –ü—Ä–∏–¥—É–º–∞–π—Ç–µ —Ç—Ä–∏ –æ–ø—Ä–∞–≤–¥–∞–Ω–∏—è.",
        "choices": {
            "1": "excuse_1",
            "2": "excuse_2",
            "3": "excuse_3"
        }
    },
    "excuse_1": {
        "title": "–ü–µ—Ä–≤–æ–µ –æ–ø—Ä–∞–≤–¥–∞–Ω–∏–µ",
        "description": "–í–≤–µ–¥–∏—Ç–µ –ø–µ—Ä–≤–æ–µ –æ–ø—Ä–∞–≤–¥–∞–Ω–∏–µ.",
        "choices": {}
    },
    "excuse_2": {
        "title": "–í—Ç–æ—Ä–æ–µ –æ–ø—Ä–∞–≤–¥–∞–Ω–∏–µ",
        "description": "–í–≤–µ–¥–∏—Ç–µ –≤—Ç–æ—Ä–æ–µ –æ–ø—Ä–∞–≤–¥–∞–Ω–∏–µ.",
        "choices": {}
    },
    "excuse_3": {
        "title": "–¢—Ä–µ—Ç—å–µ –æ–ø—Ä–∞–≤–¥–∞–Ω–∏–µ",
        "description": "–í–≤–µ–¥–∏—Ç–µ —Ç—Ä–µ—Ç—å–µ –æ–ø—Ä–∞–≤–¥–∞–Ω–∏–µ.",
        "choices": {
            "1": "girlfriend_agrees"
        }
    },
    "girlfriend_agrees": {
        "title": "–û–Ω–∞ —Å–æ–≥–ª–∞—à–∞–µ—Ç—Å—è",
        "description": "–§—É—Ö! –û–Ω–∞ –Ω–∞–∫—Ä—É—Ç–∏–ª–∞ —Å–µ–±—è –Ω–∞ –Ω–µ—Ä–≤–∞—Ö... –ö–æ–Ω–µ—á–Ω–æ —Å–æ–≥–ª–∞—Å–Ω–∞.",
        "choices": {
            "1": "final_agreement"
        }
    },
    "call_martin": {
        "title": "–ü–æ–∑–≤–æ–Ω–∏—Ç—å –ú–∞—Ä—Ç–∏–Ω—É",
        "description": "–í—ã –ø–æ–∑–≤–æ–Ω–∏–ª–∏ –≤–∞—à–µ–º—É –¥—Ä—É–≥—É, –∑–∞–∫–ª—è—Ç–æ–º—É –≤—Ä–∞–≥—É –±–∏–∑–Ω–µ—Å–º–µ–Ω–∞, –∏ –æ–Ω –≤–∞—Å —Ö–æ—á–µ—Ç —É–±–∏—Ç—å.",
        "choices": {
            "1": "final_attack"
        }
    },
    "call_louise": {
        "title": "–ü–æ–∑–≤–æ–Ω–∏—Ç—å –õ—É–∏–∑–µ",
        "description": "–í—ã –∑–≤–æ–Ω–∏—Ç–µ –≤–µ—Ä–Ω–æ–º—É –¥—Ä—É–≥—É –∏ –ø–æ–ª—É—á–∞–µ—Ç–µ –¥–æ–≤–µ—Ä–∏–µ –±–∏–∑–Ω–µ—Å–º–µ–Ω–∞. –û–Ω –¥–∞—Ä–∏—Ç –≤–∞–º –∫–≤–∞—Ä—Ç–∏—Ä—É.",
        "choices": {
            "1": "final_apartment"
        }
    },
    "call_winnie": {
        "title": "–ü–æ–∑–≤–æ–Ω–∏—Ç—å –í–∏–Ω–Ω–∏",
        "description": "–í—ã –ø–æ–∑–≤–æ–Ω–∏–ª–∏ –≤–∞—à–µ–º—É –¥—Ä—É–≥—É, –∑–∞–∫–ª—è—Ç–æ–º—É –≤—Ä–∞–≥—É –±–∏–∑–Ω–µ—Å–º–µ–Ω–∞, –∏ –æ–Ω –≤–∞—Å —Ö–æ—á–µ—Ç —É–±–∏—Ç—å.",
        "choices": {
            "1": "final_attack"
        }
    },
    "call_mike": {
        "title": "–ü–æ–∑–≤–æ–Ω–∏—Ç—å –ú–∞–π–∫—É",
        "description": "–í—ã –∑–≤–æ–Ω–∏—Ç–µ –≤–µ—Ä–Ω–æ–º—É –¥—Ä—É–≥—É –∏ –ø–æ–ª—É—á–∞–µ—Ç–µ –¥–æ–≤–µ—Ä–∏–µ –±–∏–∑–Ω–µ—Å–º–µ–Ω–∞. –û–Ω –¥–∞—Ä–∏—Ç –≤–∞–º –∫–≤–∞—Ä—Ç–∏—Ä—É.",
        "choices": {
            "1": "final_apartment"
        }
    },
    "final_arrest": {
        "title": "–í–∞—Å –∞—Ä–µ—Å—Ç–æ–≤—ã–≤–∞—é—Ç",
        "description": "–ö–û–ù–¶–û–í–ö–ê –Ω–æ–º–µ—Ä 1: –í–∞—Å –∞—Ä–µ—Å—Ç–æ–≤–∞–ª–∏ –∑–∞ —É—á–∞—Å—Ç–∏–µ –≤ –¥—Ä–∞–∫–µ. –≠—Ç–æ –∫–æ–Ω–µ—Ü –≤–∞—à–µ–π –∏—Å—Ç–æ—Ä–∏–∏.",
        "choices": {}
    },
    "final_alone": {
        "title": "–û—Å—Ç–∞–≤–∞—Ç—å—Å—è –æ–¥–Ω–æ–º—É",
        "description": "–ö–û–ù–¶–û–í–ö–ê –Ω–æ–º–µ—Ä 2: –í–∞—Å –±—Ä–æ—Å–∞–µ—Ç –¥–µ–≤—É—à–∫–∞, –∏ –≤—ã –ø—Ä–æ–≤–æ–¥–∏—Ç–µ –æ—Ç–ø—É—Å–∫ –æ–¥–∏–Ω. –≠—Ç–æ –∫–æ–Ω–µ—Ü –≤–∞—à–µ–π –∏—Å—Ç–æ—Ä–∏–∏.",
        "choices": {}
    },
    "final_no_proposal": {
        "title": "–ù–µ—Ç –ø—Ä–µ–¥–ª–æ–∂–µ–Ω–∏—è",
        "description": "–ö–û–ù–¶–û–í–ö–ê –Ω–æ–º–µ—Ä 3: –û–Ω–∞ —Ç–∞–∫ –¥–æ–ª–≥–æ –∂–¥–∞–ª–∞ –≤–∞—à–µ–≥–æ –ø—Ä–µ–¥–ª–æ–∂–µ–Ω–∏—è, —É–≤—ã, —ç—Ç–æ –±—ã–ª –ø–æ—Å–ª–µ–¥–Ω–∏–π —à–∞–Ω—Å. –≠—Ç–æ –∫–æ–Ω–µ—Ü –≤–∞—à–µ–π –∏—Å—Ç–æ—Ä–∏–∏.",
        "choices": {}
    },
    "final_agreement": {
        "title": "–°–æ–≥–ª–∞—Å–∏–µ",
        "description": "–ö–û–ù–¶–û–í–ö–ê –Ω–æ–º–µ—Ä 4: –û–Ω–∞ —Å–æ–≥–ª–∞—Å–∏–ª–∞—Å—å —Å –ø—Ä–µ–¥–ª–æ–∂–µ–Ω–∏–µ–º –∏ –≤—ã —Å—á–∞—Å—Ç–ª–∏–≤—ã –≤–º–µ—Å—Ç–µ. –≠—Ç–æ –∫–æ–Ω–µ—Ü –≤–∞—à–µ–π –∏—Å—Ç–æ—Ä–∏–∏.",
        "choices": {}
    },
    "final_attack": {
        "title": "–ê—Ç–∞–∫–∞",
        "description": "–ö–û–ù–¶–û–í–ö–ê –Ω–æ–º–µ—Ä 5: –í–∞—à –∑–∞–∫–ª—è—Ç—ã–π –≤—Ä–∞–≥ –∞—Ç–∞–∫—É–µ—Ç –≤–∞—Å, –∏ –≤—ã –≤—ã–Ω—É–∂–¥–µ–Ω—ã –±–æ—Ä–æ—Ç—å—Å—è –∑–∞ —Å–≤–æ—é –∂–∏–∑–Ω—å. –≠—Ç–æ –∫–æ–Ω–µ—Ü –≤–∞—à–µ–π –∏—Å—Ç–æ—Ä–∏–∏.",
        "choices": {}
    },
    "final_apartment": {
        "title": "–ü–æ–ª—É—á–µ–Ω–∏–µ –∫–≤–∞—Ä—Ç–∏—Ä—ã",
        "description": "–ö–û–ù–¶–û–í–ö–ê –Ω–æ–º–µ—Ä 6: –í—ã –ø–æ–ª—É—á–∞–µ—Ç–µ –¥–æ–≤–µ—Ä–∏–µ –±–∏–∑–Ω–µ—Å–º–µ–Ω–∞, –∏ –æ–Ω –¥–∞—Ä–∏—Ç –≤–∞–º –∫–≤–∞—Ä—Ç–∏—Ä—É. –≠—Ç–æ –∫–æ–Ω–µ—Ü –≤–∞—à–µ–π –∏—Å—Ç–æ—Ä–∏–∏.",
        "choices": {}
    }
}


contacts = ["–ú–∞—Ä—Ç–∏–Ω", "–õ—É–∏–∑–∞", "–í–∏–Ω–Ω–∏", "–ú–∞–π–∫"]  # –°–ø–∏—Å–æ–∫ –∫–æ–Ω—Ç–∞–∫—Ç–æ–≤

# –§—É–Ω–∫—Ü–∏—è –¥–ª—è –æ—Ç–æ–±—Ä–∞–∂–µ–Ω–∏—è —Ç–µ–∫—É—â–µ–π —Å—Ü–µ–Ω—ã –∏ –ø–æ–ª—É—á–µ–Ω–∏—è –≤—ã–±–æ—Ä–∞ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è
def display_scene(scene):
    print(f"\n–û–ø–∏—Å–∞–Ω–∏–µüë®üèΩ‚Äçüíª:\n {stories[scene]['description']}")
    if scene == "boat_dinner":
        return handle_boat_dinner()
    elif scene == "accept_invitation":
        return handle_accept_invitation()
    elif scene == "propose":
        return handle_propose()

    choices = stories[scene]['choices']
    if len(choices) == 1:
        next_scene = list(choices.values())[0]
        print(f"–ê–≤—Ç–æ–º–∞—Ç–∏—á–µ—Å–∫–∏–π –≤—ã–±–æ—Ä: {stories[next_scene]['title']}")
        return next_scene
    elif choices:
        for choice, next_scene in choices.items():
            print(f"{choice}. {stories[next_scene]['title']}")
        user_choice = input("–í—ã–±–µ—Ä–∏—Ç–µ –≤–∞—Ä–∏–∞–Ω—Ç: ")
        if user_choice in choices:
            return choices[user_choice]
        else:
            print("–ù–µ–ø—Ä–∞–≤–∏–ª—å–Ω—ã–π –≤—ã–±–æ—Ä. –ü–æ–ø—Ä–æ–±—É–π—Ç–µ –µ—â–µ —Ä–∞–∑.")
            return display_scene(scene)
    else:
        return None

# –§—É–Ω–∫—Ü–∏—è –¥–ª—è –æ–±—Ä–∞–±–æ—Ç–∫–∏ —Å—Ü–µ–Ω—ã "–£—Å—Ç—Ä–æ–∏—Ç—å —É–∂–∏–Ω –Ω–∞ –ª–æ–¥–∫–µ"
def handle_boat_dinner():
    correct_code = "1205"  # –ö–æ–¥, –∫–æ—Ç–æ—Ä—ã–π –Ω—É–∂–Ω–æ —É–≥–∞–¥–∞—Ç—å
    attempt = ""
    while attempt != correct_code:
        attempt = input("–û–Ω –æ–ø–ª–∞—á–∏–≤–∞–µ—Ç –∫–∞—Ä—Ç–æ–π. –ó–∞–±—ã–ª –ø–∞—Ä–æ–ª—å, –ø–æ–¥–±–∏—Ä–∞–µ—Ç –∫–æ–¥: ")
        if attempt == correct_code:
            print("–í—ã –≤—Å–ø–æ–º–Ω–∏–ª–∏ –∫–æ–¥ –∏ –ø—Ä–æ—à–ª–∏ –Ω–∞ —É–∂–∏–Ω.")
        else:
            print("–ù–µ–ø—Ä–∞–≤–∏–ª—å–Ω—ã–π –∫–æ–¥. –ü–æ–ø—Ä–æ–±—É–π—Ç–µ –µ—â–µ —Ä–∞–∑.")

# –§—É–Ω–∫—Ü–∏—è –¥–ª—è –æ–±—Ä–∞–±–æ—Ç–∫–∏ —Å—Ü–µ–Ω—ã "–ü—Ä–∏–Ω—è—Ç—å –ø—Ä–∏–≥–ª–∞—à–µ–Ω–∏–µ"
def handle_accept_invitation():
    print("–í—ã–±–∏—Ä–∞–µ—Ç–µ, –∫–æ–º—É –ø–æ–∑–≤–æ–Ω–∏—Ç—å –∏–∑ —Å–ø–∏—Å–∫–∞ –∫–æ–Ω—Ç–∞–∫—Ç–æ–≤:")
    for i, contact in enumerate(contacts, 1):
        print(f"{i}. {contact}")
    user_choice = input("–í–≤–µ–¥–∏—Ç–µ –Ω–æ–º–µ—Ä –∫–æ–Ω—Ç–∞–∫—Ç–∞üë®üèΩ‚Äçüíª: ")
    if user_choice == "1":
        return "call_martin"
    elif user_choice == "2":
        return "call_louise"
    elif user_choice == "3":
        return "call_winnie"
    elif user_choice == "4":
        return "call_mike"
    else:
        print("–ù–µ–ø—Ä–∞–≤–∏–ª—å–Ω—ã–π –≤—ã–±–æ—Ä. –ü–æ–ø—Ä–æ–±—É–π—Ç–µ –µ—â–µ —Ä–∞–∑.")
        return

# –§—É–Ω–∫—Ü–∏—è –¥–ª—è –æ–±—Ä–∞–±–æ—Ç–∫–∏ —Å—Ü–µ–Ω—ã "–°–¥–µ–ª–∞—Ç—å –ø—Ä–µ–¥–ª–æ–∂–µ–Ω–∏–µ"
def handle_propose():
    print("–û –Ω–µ—Ç! –û–Ω–∞ –ø–æ–¥–æ–∑—Ä–µ–≤–∞–µ—Ç –≤–∞—Å –≤ –Ω–µ–≤–µ—Ä–Ω–æ—Å—Ç–∏, –≤–∞–º –Ω—É–∂–Ω–æ –æ–ø—Ä–∞–≤–¥–∞—Ç—å—Å—è. –í–≤–µ–¥–∏—Ç–µ 3 –ø—Ä–∏—á–∏–Ω—ã, –ø–æ—á–µ–º—É –≤—ã —Ö–æ–¥–∏–ª–∏ –≤ —Ä–µ—Å—Ç–æ—Ä–∞–Ω —Å –¥—Ä—É–≥–æ–π –¥–µ–≤—É—à–∫–æ–π.")
    excuses = {}
    excuses[1] = input("–í–≤–µ–¥–∏—Ç–µ –ø–µ—Ä–≤–æ–µ –æ–ø—Ä–∞–≤–¥–∞–Ω–∏–µüëç: ")
    excuses[2] = input("–í–≤–µ–¥–∏—Ç–µ –≤—Ç–æ—Ä–æ–µ –æ–ø—Ä–∞–≤–¥–∞–Ω–∏–µüëç: ")
    excuses[3] = input("–í–≤–µ–¥–∏—Ç–µ —Ç—Ä–µ—Ç—å–µ –æ–ø—Ä–∞–≤–¥–∞–Ω–∏–µüëç: ")

    # –ó–∞–ø–∏—Å—å –æ–ø—Ä–∞–≤–¥–∞–Ω–∏–π –≤ —Ñ–∞–π–ª
    with open("excuses.txt", "w") as file:
        for i in range(1, 4):
            file.write(f"–û–ø—Ä–∞–≤–¥–∞–Ω–∏–µ {i}: {excuses[i]}\n")

    return "girlfriend_agrees"

# –û—Å–Ω–æ–≤–Ω–∞—è —Ñ—É–Ω–∫—Ü–∏—è
def start_story():
    current_scene = "start"
    while current_scene:
        current_scene = display_scene(current_scene)

# –ó–∞–ø—É—Å–∫ –∏—Å—Ç–æ—Ä–∏–∏
start_story()
```
